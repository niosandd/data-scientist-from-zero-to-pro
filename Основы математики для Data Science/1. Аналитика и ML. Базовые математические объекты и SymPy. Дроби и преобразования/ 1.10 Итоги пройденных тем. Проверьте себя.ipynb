```python
from sympy import *
```

```python
# Задание № 1
```

```python
x, y = symbols('x, y')
```

```python
expression = (2*x + 3*y)**2 - 3*x * (4*x/3 + 4*y)
expression
```

```python
expression = simplify(expression)
expression
```

```python
expression.subs({x: 1.038, y: sqrt(3)})
```
