```python
def ln_sum(n: int):
    """
    Функция  возвращает  кортеж  из значения и 
    выражения суммы логарифмов чисел от 1 до n
    """
    expr = UnevaluatedExpr(log(1))
    for i in range(2, n + 1):
        expr += UnevaluatedExpr(log(i))
    return expr.simplify().evalf()
```

```python
ln_sum_result = ln_sum(10)
ln_sum_result
```

```python
def product_ln(n: int):
    """
    Функция возвращает кортеж из значения и выра-
    жения логарифма произведения  чисел от 1 до n
    """
    expr = UnevaluatedExpr(1)
    for i in range(2, n + 1):
        expr *= UnevaluatedExpr(i)
    expr = log(expr)
    return expr.simplify().evalf()
```

```python
product_ln_result = product_ln(10)
product_ln_result
```
