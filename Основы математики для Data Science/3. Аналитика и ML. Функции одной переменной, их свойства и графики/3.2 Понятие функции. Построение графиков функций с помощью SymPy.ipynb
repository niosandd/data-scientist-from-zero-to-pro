```python
# 3.2 Понятие функции. Построение графиков функций с помощью SymPy
```

```python
from sympy import *
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol("x")
```

```python
f = x + 1
f
```

```python
f.subs(x, 0)
```

```python
f.subs(x, -1)
```

```python
plot(f)
```

```python
f = 3 * x + 5
f
```

```python
f.subs(x, 5/3)
f.subs(x, 2)
plot(f)
```

```python
f = 3 * x **2 + 3 * x - 6
f.subs(x, 0.5)
f.subs(x, 7)
plot(f)
```

```python
f = 2 * x ** 4 + x ** 3 - 6 * x ** 2 + 7 * x + 1
f.subs(x, 3)
f.subs(x, -5)
plot(f)
```

```python
f = (x**2)**x
f.subs(x, 0)
f.subs(x, 10)
plot(f)
```
