```python
# линейный поиск
tmp_list = [1, 2, 3, 4, 5, 6, 7, 17, 8, 9, 10]
number = 17
```

```python
for i in range(len(tmp_list)):
    if tmp_list[i] == number:
        print(i)
        break
```

```python
# бинарный поиск
number = 17
left = 0
right = len(tmp_list) - 1
middle = 0
tmp_list = sorted(tmp_list)

while left <= right:
    middle = (left + right) // 2
    
    if number > tmp_list[middle]:
        left = middle + 1
    elif number < tmp_list[middle]:
        right = middle - 1
    else: 
        print(middle)
        break
```

```python
import math
tmp_list = [1, 2, 3, 4, 5, 6, 7, 17, 8, 9, 10]
number = 17

tmp_list = sorted(tmp_list)
prev = 0
n = len(tmp_list)
step = int(math.sqrt(n))

while prev < n and tmp_list[min(step, n) - 1] < number:
    prev = step
    step += int(math.sqrt(n))

while prev < min(step, n):
    if tmp_list[prev] == number:
        print(prev)
        break
    prev += 1

else: 
    print('Ничего не найдено')
```

```python
tmp_list = [1, 2, 3, 4, 5, 6, 7, 17, 8, 9, 10]
number = 8
# O(n)
for i in range(len(tmp_list)):
    if tmp_list[i] == number:
        print(i)
```

```python
# O(log(n))
tmp_list = sorted(tmp_list)
left = 0
right = len(tmp_list) - 1
middle = 0
while left <= right:
    middle = (left + right) // 2
    
    if tmp_list[middle] < number:
        left = middle + 1
    elif tmp_list[middle] > number:
        right = middle - 1
    else:
        print(middle)
        break
```
