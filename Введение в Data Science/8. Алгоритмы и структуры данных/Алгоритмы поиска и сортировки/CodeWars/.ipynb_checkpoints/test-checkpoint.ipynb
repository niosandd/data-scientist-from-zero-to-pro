```python
def validate_pin(pin):
    if len(str(pin)) == 4 or len(str(pin)) == 6 and isnumeric(pin):
        return True
    else:
        return False
```

```python
def opposite(number):
    if number > 0:
        print(number * (-1))
    else:
        print(abs(number))
```

```python
def check(seq, elem):
    for i in range(len(seq)):
        if seq[i] == elem:
            return True
    return False
```

```python
seq = [1,2,3,4,5,6,7, -1, 22, -66]
check(seq, -6)
```

```python
def array_diff(a, b):
    result = []
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i] != b[j]:
                result.append(b[j\])
    return result
```

```python
a = [1, 2, 3]
b = [1, 2]
```

```python
def descending_order(num):
    num = list(str(num))
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(num) - 1):
            if num[int(i)] < num[int(i) + 1]:
                num[int(i)], num[int(i) + 1] = num[int(i) + 1], num[int(i)]
                swapped = True  
    return ''.join(num)       
```

```python
descending_order(123456789)
```
