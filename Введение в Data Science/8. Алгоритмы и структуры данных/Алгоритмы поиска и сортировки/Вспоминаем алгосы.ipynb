```python
def QuickSort(nlist):
    if len(nlist) <= 1:
        return nlist
    
    pivot = nlist[len(nlist) // 2]
    left = [x for x in nlist if x < pivot]
    middle = [x for x in nlist if x == pivot]
    right = [x for x in nlist if x > pivot]
    return QuickSort(left) + middle + QuickSort(right)
```
