# Линейный поиск

```python
def LinearSearch(numbers_list, number_for_search):
    iterations = 0
    for i in range(len(numbers_list)):
        iterations += 1
        if numbers_list[i] == number_for_search:
            print(i)
            print(iterations)
```

```python
LinearSearch(numbers_list, 22)
```

Простыми словами:

Например, если у нас есть список list = [1, 2, 3, 4]

То, для каждого индекса из длины(len) равного 4 списка list проверяем условие, где если значение индекса равняется тому числу, которое мы ищем, то выводим его индекс расположение.

Личный тупик был в том, что, если мы смотрим на массив, то че это за numbers_list[i], а numbers_list[i] это когда выводит значение индексовское, типа если мы берем list[0], то получаем 1, если list[1], то 2 и т.д.

# Бинарный поиск

```python
numbers_list = [7, 88, 2, 7, 22, 7, 3, 1, 66, 1]
# [1, 1, 2, 3, 7, 7, 7, 22, 66, 88] - sorted
```

```python
def BinarySearch(numbers_list, number_for_search):
    iterations = 0
    numbers_list = sorted(numbers_list)
    left = 0
    right = len(numbers_list) - 1
    middle = 0
    while left <= right:
        iterations += 1
        middle = (left+right) // 2
        if number_for_search > numbers_list[middle]:
            left = middle + 1
        elif number_for_search < numbers_list[middle]:
            right= middle - 1
        else:
            print('Индексом числа является число:', middle)
            break
    print()        
    print('Для такого количества значений:', len(numbers_list))
    print('Потребовалось всего итераций:',iterations)
```

```python
BinarySearch(numbers_list, 22)
```
