# LinearSearch

```python
# O(n)
def LinearSearch(numbers_list, number_for_search):
    iterations = 0
    for i in range(len(numbers_list)):
        iterations += 1
        if numbers_list[i] == number_for_search:
            print('Индексом числа является зачение:', i)
            print()
            print('Всего значений:', len(numbers_list))
            print('Потребовалось всего итераций:',iterations)
```

```python
LinearSearch(numbers_list, 22)
```

То, для каждого индекса из длины(len) равного 4 списка list проверяем условие, где если значение индекса равняется тому числу, которое мы ищем, то выводим его индекс расположение.

# BinarySearch

```python
numbers_list = [7, 88, 2, 7, 22, 7, 3, 1, 66, 1]
# [1, 1, 2, 3, 7, 7, 7, 22, 66, 88] - sorted
```

```python
# O(log(n))
def BinarySearch(numbers_list, number_for_search):
    iterations = 0
    numbers_list = sorted(numbers_list)
    left = 0
    right = len(numbers_list) - 1
    middle = 0
    while left <= right:
        iterations += 1
        middle = (left+right) // 2
        if number_for_search > numbers_list[middle]:
            left = middle + 1
        elif number_for_search < numbers_list[middle]:
            right= middle - 1
        else:
            print('Индексом числа является зачение:', middle)
            break
    print()        
    print('Всего значений:', len(numbers_list))
    print('Потребовалось всего итераций:',iterations)
```

```python
BinarySearch(numbers_list, 22)
```

```python
def BinarySearch_lower_bound(numbers_list, number_for_search):
    iterations = 0
    left = 0
    right = len(numbers_list)          # здесь специально len, а не len-1
    while left < right:
        iterations += 1
        middle = (left + right) // 2
        if number_for_search > numbers_list[middle]:
            left = middle + 1
        else:
            right = middle
    print('Потребовалось итераций (lower_bound):', iterations)
    return left
```

```python
BinarySearch_lower_bound(numbers_list, 7)
```

```python
def BinarySearch_upper_bound(numbers_list, number_for_search):
    iterations = 0
    left = 0
    right = len(numbers_list)          # здесь специально len, а не len-1
    while left < right:
        iterations += 1
        middle = (left + right) // 2
        if number_for_search >= numbers_list[middle]:
            left = middle + 1
        else:
            right = middle
    print('Потребовалось итераций (upper_bound):', iterations)
    return left
```

```python
BinarySearch_upper_bound(numbers_list, 7)
```

# JumpSearch

```python
import math
numbers_list = [15, 88, 3, 11, 22, 9, 5, 7, 66, 1]
number = 66
```

```python
# O(sqrt(n))
def JumpSearch(numbers_list, number_for_search):
    iterations = 0
    numbers_list = sorted(numbers_list)
    # [1, 1, 2, 3, 7, 7, 7, 22, 66, 88] - sorted
    n = len(numbers_list)
    step = int(math.sqrt(n))
    prev = 0

    # 1. Прыжки вперёд
    while prev < n and numbers_list[min(step, n) - 1] < number_for_search:
        iterations += 1
        prev = step
        step += int(math.sqrt(n))

    # 2. Линейный поиск внутри блока [prev, min(step, n))
    while prev < min(step, n):
        iterations += 1
        if numbers_list[prev] == number_for_search:
            print('Индексом числа является зачение:', prev)
            break
        prev += 1
    else:
        print('Элемент не найден')

    print()
    print('Всего значений:', len(numbers_list))
    print('Потребовалось всего итераций:', iterations)
```

```python
JumpSearch(numbers_list, number)
```

# FibonacciSearch (Не спрашивают на собеседованиях)

```python
# O(log(n))
def FibonacciSearch(numbers_list, number_for_search):
    iterations = 0
    n = len(numbers_list) # 10
    numbers_list = sorted(numbers_list)
    # Находим наименьшее число Фибоначчи >= длины списка
    fib_m_minus_2 = 0
    fib_m_minus_1 = 1
    fib_m = fib_m_minus_1 + fib_m_minus_2 # 1
    
    while fib_m < n: # 10
        fib_m_minus_2 = fib_m_minus_1 #  1 1 2 3 5
        fib_m_minus_1 = fib_m # 1 2 3 5 8
        fib_m = fib_m_minus_1 + fib_m_minus_2 # 2 3 5 8 13 
    
    index = -1
    
    while fib_m > 1: # 13 > 1
        iterations += 1
        i = min(index + fib_m_minus_2, n - 1) # 4, 9
        
        if number_for_search > numbers_list[i]:
            fib_m = fib_m_minus_1
            fib_m_minus_1 = fib_m_minus_2
            fib_m_minus_2 = fib_m - fib_m_minus_1
            index = i
        elif number_for_search < numbers_list[i]:
            fib_m = fib_m_minus_2
            fib_m_minus_1 = fib_m_minus_1 - fib_m_minus_2
            fib_m_minus_2 = fib_m - fib_m_minus_1
        else:
            print('Индекс найденного числа является зачение:', i)
            print('Потребовалось итераций:', iterations)
            return i
    
    # Проверка последнего элемента
    if fib_m_minus_1 and index < n - 1 and numbers_list[index + 1] == number_for_search:
        print('Индекс найденного числа является зачение:', index + 1)
        print('Потребовалось итераций:', iterations)
        return index + 1
    
    print('Число не найдено')
    print('Потребовалось итераций:', iterations)
    return -1
```

```python
numbers_list = [15, 88, 3, 11, 22, 9, 5, 7, 66, 1]
number = 66
```

```python
FibonacciSearch(numbers_list, 66)
```

# ExponentialSearch

```python
# O(log i)
def ExponentialSearch(numbers_list, number_for_search):
    numbers_list = sorted(numbers_list)
    if not numbers_list:
        print("Список пустой")
        return None

    if numbers_list[0] == number_for_search:
        print('Индексом числа является зачение:', 0)
        return 0

    # Экспоненциально находим правую границу диапазона
    bound = 1 
    while bound < len(numbers_list) and numbers_list[bound] <= number_for_search:
        bound *= 2
    sub_list = numbers_list[:min(bound, len(numbers_list))]
    return BinarySearch(sub_list, number_for_search)
```

```python
ExponentialSearch(numbers_list, 22)
```

# InterpolationSearch

```python
# O(log log n)
def InterpolationSearch(numbers_list, number_for_search):
    numbers_list = sorted(numbers_list)
    
    if not numbers_list:
        print("Список пустой")
        return -1

    iterations = 0
    left = 0
    right = len(numbers_list) - 1

    while (left <= right and
           number_for_search >= numbers_list[left] and
           number_for_search <= numbers_list[right]):
        
        iterations += 1

        # Защита от деления на ноль
        if numbers_list[right] == numbers_list[left]:
            break

        # Формула интерполяции
        pos = left + int(
            ((right - left) / (numbers_list[right] - numbers_list[left]))
            * (number_for_search - numbers_list[left])
        )

        if numbers_list[pos] == number_for_search:
            print('Индексом числа является значение:', pos)
            print()
            print('Всего значений:', len(numbers_list))
            print('Потребовалось всего итераций:', iterations)
            return pos

        elif numbers_list[pos] < number_for_search:
            left = pos + 1
        else:
            right = pos - 1

    # Если элемент не найден
    print('Элемент не найден')
    print()
    print('Всего значений:', len(numbers_list))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
InterpolationSearch(numbers_list, 66)
```
