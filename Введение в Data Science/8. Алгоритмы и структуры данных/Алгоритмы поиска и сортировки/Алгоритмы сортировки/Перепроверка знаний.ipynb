```python
tmp_list = [1, 2, 3, 6, 7, 8, 2, 1, 11, 55, 33, 22]
```

```python
def BubbleSort(tmp_list):
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(tmp_list) - 1):
            if tmp_list[i] > tmp_list[i+1]:
                tmp_list[i], tmp_list[i+1] = tmp_list[i+1], tmp_list[i]
                swapped = True
```

```python
BubbleSort(tmp_list)
```
