## 1. Что такое поиск

Поиск — это задача определения наличия элемента в наборе данных и, при необходимости, нахождения его позиции.  

На небольших объёмах данных способ поиска почти не имеет значения. Однако при росте размера входных данных эффективность выбранного метода начинает напрямую влиять на время работы программы. Именно поэтому существуют различные алгоритмы поиска, каждый из которых представляет собой компромисс между скоростью, требованиями к предварительной обработке данных и сложностью реализации.

---

## 2. Оценка эффективности: нотация Big-O

Чтобы сравнивать алгоритмы независимо от конкретного языка программирования и мощности компьютера, используют асимптотическую оценку сложности — нотацию Big-O.

Big-O показывает, как растёт время выполнения алгоритма при увеличении размера входных данных \( n \). При этом отбрасываются константные множители и младшие члены. Важны только доминирующие тенденции.

Основные классы сложности:

| Обозначение | Название                  | Характер роста                          |
|-------------|---------------------------|-----------------------------------------|
| \( O(1) \)  | Константная               | Время почти не зависит от \( n \)       |
| \( O(\log n) \) | Логарифмическая        | Рост крайне медленный                   |
| \( O(\sqrt{n}) \) | Квадратный корень    | Умеренный рост                          |
| \( O(n) \)  | Линейная                  | Время растёт пропорционально \( n \)    |
| \( O(n \log n) \) | Линейно-логарифмическая | Быстрее линейной, но медленнее квадратичной |
| \( O(n^2) \) | Квадратичная             | Быстрый рост                            |

Оценка обычно даётся для худшего случая. На практике также рассматривают средний и лучший случаи.

---

## 3. Линейный поиск — \( O(n) \)

Самый простой метод. Алгоритм последовательно просматривает каждый элемент массива до тех пор, пока не найдёт искомый или не достигнет конца.

Преимущества:
- не требует предварительной сортировки;
- прост в реализации.

Недостаток — в худшем случае приходится просмотреть все элементы. Поэтому на больших объёмах данных линейный поиск становится неэффективным.

---

## 4. Бинарный поиск — \( O(\log n) \)

Ключевое условие: массив должен быть отсортирован.

Алгоритм на каждом шаге берёт элемент в середине текущего диапазона и сравнивает его с искомым значением. В зависимости от результата сравнения одна из половин диапазона полностью отбрасывается. Таким образом, область поиска сокращается вдвое после каждого сравнения.

Благодаря этому свойству бинарный поиск работает значительно быстрее линейного. Для массива из миллиона элементов достаточно примерно двадцати сравнений.

На основе бинарного поиска строятся две важные модификации:

- **lower_bound** — находит первую позицию, в которую можно вставить элемент (первый элемент, больший или равный искомому);
- **upper_bound** — находит первую позицию после всех элементов, равных искомому (первый элемент, строго больший искомого).

Эти варианты особенно полезны при работе с массивами, содержащими дубликаты.

---

## 5. Прыжковый поиск (Jump Search) — \( O(\sqrt{n}) \)

Метод занимает промежуточное положение между линейным и бинарным поиском.

Массив просматривается прыжками фиксированной длины (обычно \( \sqrt{n} \)). Как только прыжок «перескакивает» искомое значение, выполняется линейный поиск внутри предыдущего блока.

Выбор размера прыжка \( \sqrt{n} \) является оптимальным: он минимизирует суммарное количество операций в худшем случае.

---

## 6. Экспоненциальный поиск — \( O(\log i) \)

Алгоритм состоит из двух этапов. Сначала экспоненциально расширяется диапазон поиска (позиции 1, 2, 4, 8, 16…), пока не будет найдена верхняя граница, превосходящая искомое значение. Затем в найденном диапазоне выполняется обычный бинарный поиск.

Сложность зависит от позиции элемента \( i \), поэтому её записывают как \( O(\log i) \). Метод особенно эффективен, когда искомый элемент расположен близко к началу массива или когда размер массива заранее неизвестен.

---

## 7. Интерполяционный поиск — \( O(\log \log n) \)

Этот алгоритм является развитием идеи бинарного поиска. Вместо деления диапазона строго пополам он пытается оценить вероятную позицию элемента на основе его значения, используя формулу линейной интерполяции.

На данных с относительно равномерным распределением интерполяционный поиск работает чрезвычайно быстро и достигает сложности \( O(\log \log n) \). Однако при сильно неравномерном распределении значений его эффективность может значительно снизиться.

---

## 8. Поиск Фибоначчи — \( O(\log n) \)

Алгоритм использует числа Фибоначчи вместо деления диапазона пополам. Это позволяет выполнять сужение области поиска только с помощью операций сложения и вычитания.

С точки зрения асимптотической сложности поиск Фибоначчи эквивалентен бинарному. Исторически он был полезен на архитектурах, где операция деления выполнялась медленно. В современных условиях практическое преимущество практически отсутствует, однако метод представляет теоретический интерес.

---

## 9. Общая картина

Можно выстроить алгоритмы по возрастанию эффективности на больших объёмах данных:

1. Линейный поиск — универсальный, но медленный.
2. Прыжковый поиск — ускорение за счёт пропуска элементов.
3. Бинарный поиск и поиск Фибоначчи — радикальное сокращение области поиска.
4. Экспоненциальный поиск — быстрое определение примерного диапазона с последующим бинарным поиском.
5. Интерполяционный поиск — наиболее быстрый на равномерно распределённых данных.

Выбор конкретного алгоритма определяется характером данных, частотой их изменения и допустимой стоимостью предварительной сортировки.

# Алгоритмы поиска

# LinearSearch

```python
# ============================================
# O(n) — Линейный поиск
# ============================================
def LinearSearch(numbers_list, number_for_search):
    iterations = 0
    arr = numbers_list.copy()          # не трогаем оригинал

    for i in range(len(arr)):
        iterations += 1
        if arr[i] == number_for_search:
            print('Индекс найденного числа:', i)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return i

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
n = np.linspace(1, 200, 500)
y = n

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#c0392b", linewidth=2.8, label="O(n)")

plt.title("Линейный поиск — O(n)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 200)

plt.tight_layout()
plt.show()
```

# BinarySearch

```python
# ============================================
# O(log n) — Бинарный поиск
# ============================================
def BinarySearch(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)

    left = 0
    right = len(arr) - 1

    while left <= right:
        iterations += 1
        middle = (left + right) // 2

        if number_for_search > arr[middle]:
            left = middle + 1
        elif number_for_search < arr[middle]:
            right = middle - 1
        else:
            print('Индекс найденного числа:', middle)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return middle

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
n = np.linspace(1, 200, 500)
y = np.log2(n)

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#27ae60", linewidth=2.8, label="O(log n)")

plt.title("Бинарный поиск — O(log n)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 8)

plt.tight_layout()
plt.show()
```

## Модификаторы к BinarySearch 

```python
# ============================================
# Нижняя граница (lower_bound)
# Первый элемент, который >= number_for_search
# ============================================
def BinarySearch_lower_bound(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)
    left = 0
    right = len(arr)

    while left < right:
        iterations += 1
        middle = (left + right) // 2

        if number_for_search > arr[middle]:
            left = middle + 1
        else:
            right = middle

    print('Нижняя граница (lower_bound):', left)
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return left
```

```python
n = np.linspace(1, 200, 500)
y = np.log2(n)

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#28b463", linewidth=2.8, label="O(log n)")

plt.title("lower_bound — O(log n)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 8)

plt.tight_layout()
plt.show()
```

```python
# ============================================
# Верхняя граница (upper_bound)
# Первый элемент, который > number_for_search
# ============================================
def BinarySearch_upper_bound(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)
    left = 0
    right = len(arr)

    while left < right:
        iterations += 1
        middle = (left + right) // 2

        if number_for_search >= arr[middle]:
            left = middle + 1
        else:
            right = middle

    print('Верхняя граница (upper_bound):', left)
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return left
```

# JumpSearch

```python
# ============================================
# O(√n) — Прыжковый поиск (Jump Search)
# ============================================
def JumpSearch(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0

    # 1. Прыжки вперёд
    while prev < n and arr[min(step, n) - 1] < number_for_search:
        iterations += 1
        prev = step
        step += int(math.sqrt(n))

    # 2. Линейный поиск внутри блока
    while prev < min(step, n):
        iterations += 1
        if arr[prev] == number_for_search:
            print('Индекс найденного числа:', prev)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return prev
        prev += 1

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
n = np.linspace(1, 200, 500)
y = np.sqrt(n)

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#e67e22", linewidth=2.8, label="O(√n)")

plt.title("Прыжковый поиск (Jump Search) — O(√n)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 15)

plt.tight_layout()
plt.show()
```

# ExponentialSearch

```python
# ============================================
# O(log i) — Экспоненциальный поиск
# ============================================
def ExponentialSearch(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)

    if not arr:
        print('Список пустой')
        print()
        print('Всего значений: 0')
        print('Потребовалось всего итераций:', iterations)
        return -1

    if arr[0] == number_for_search:
        iterations += 1
        print('Индекс найденного числа: 0')
        print()
        print('Всего значений:', len(arr))
        print('Потребовалось всего итераций:', iterations)
        return 0

    bound = 1
    while bound < len(arr) and arr[bound] <= number_for_search:
        iterations += 1
        bound *= 2
    left = bound // 2
    right = min(bound, len(arr) - 1)

    while left <= right:
        iterations += 1
        middle = (left + right) // 2

        if number_for_search > arr[middle]:
            left = middle + 1
        elif number_for_search < arr[middle]:
            right = middle - 1
        else:
            print('Индекс найденного числа:', middle)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return middle

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
n = np.linspace(1, 200, 500)
y = np.log2(n)          # в худшем случае O(log i) ≈ O(log n)

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#8e44ad", linewidth=2.8, label="O(log i) ≈ O(log n)")

plt.title("Экспоненциальный поиск — O(log i)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 8)

plt.tight_layout()
plt.show()
```

# InterpolationSearch

```python
# ============================================
# O(log log n) — Интерполяционный поиск
# ============================================
def InterpolationSearch(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)

    if not arr:
        print('Список пустой')
        print()
        print('Всего значений: 0')
        print('Потребовалось всего итераций:', iterations)
        return -1

    left = 0
    right = len(arr) - 1

    while (left <= right and
           number_for_search >= arr[left] and
           number_for_search <= arr[right]):

        iterations += 1

        # Защита от деления на ноль
        if arr[right] == arr[left]:
            break

        # Формула интерполяции
        pos = left + int(
            ((right - left) / (arr[right] - arr[left])) *
            (number_for_search - arr[left])
        )

        if arr[pos] == number_for_search:
            print('Индекс найденного числа:', pos)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return pos
        elif arr[pos] < number_for_search:
            left = pos + 1
        else:
            right = pos - 1

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```

```python
n = np.linspace(2, 200, 500)
y = np.log2(np.log2(n))

plt.figure(figsize=(9, 5.5), dpi=130)

plt.plot(n, y, color="#2980b9", linewidth=2.8, label="O(log log n)")

plt.title("Интерполяционный поиск — O(log log n)", fontsize=14, fontweight="bold", pad=12)
plt.xlabel("Количество элементов (n)", fontsize=12)
plt.ylabel("Количество операций", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.xlim(0, 200)
plt.ylim(0, 4)

plt.tight_layout()
plt.show()
```

# FibonacciSearch (Не спрашивают на собеседованиях)

```python
# ============================================
# O(log n) — Поиск Фибоначчи
# ============================================
def FibonacciSearch(numbers_list, number_for_search):
    iterations = 0
    arr = sorted(numbers_list)
    n = len(arr)

    # Находим наименьшее число Фибоначчи >= n
    fib_m_minus_2 = 0
    fib_m_minus_1 = 1
    fib_m = fib_m_minus_1 + fib_m_minus_2

    while fib_m < n:
        fib_m_minus_2 = fib_m_minus_1
        fib_m_minus_1 = fib_m
        fib_m = fib_m_minus_1 + fib_m_minus_2

    index = -1

    while fib_m > 1:
        iterations += 1
        i = min(index + fib_m_minus_2, n - 1)

        if number_for_search > arr[i]:
            fib_m = fib_m_minus_1
            fib_m_minus_1 = fib_m_minus_2
            fib_m_minus_2 = fib_m - fib_m_minus_1
            index = i
        elif number_for_search < arr[i]:
            fib_m = fib_m_minus_2
            fib_m_minus_1 = fib_m_minus_1 - fib_m_minus_2
            fib_m_minus_2 = fib_m - fib_m_minus_1
        else:
            print('Индекс найденного числа:', i)
            print()
            print('Всего значений:', len(arr))
            print('Потребовалось всего итераций:', iterations)
            return i

    # Проверка последнего элемента
    if fib_m_minus_1 and index < n - 1 and arr[index + 1] == number_for_search:
        iterations += 1
        print('Индекс найденного числа:', index + 1)
        print()
        print('Всего значений:', len(arr))
        print('Потребовалось всего итераций:', iterations)
        return index + 1

    print('Элемент не найден')
    print()
    print('Всего значений:', len(arr))
    print('Потребовалось всего итераций:', iterations)
    return -1
```
