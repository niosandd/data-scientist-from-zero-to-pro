### **Задача 1**

Дан список `list1`, который содержит целые числа. Всего в списке 12 элементов. 

Ваша задача — посчитать сумму элементов с нечётными индексами. Другими словами, нужно сложить элементы `list1[1]`, `list1[3]`, `list1[5]`, `list1[7]`, `list1[9]` и `list1[11]`. 

Чтобы решить эту задачу, используйте slicing. 

Подсказка: вам придётся указать одну границу среза и шаг.   

**Решение.** Чтобы пройтись по элементам с нечётными индексами, нужно начать с элемента с индексом 1 (указываем левую границу среза) и шагать через один элемент (указываем шаг, равный 2). Правую границу среза не указываем, поскольку перебираем все элементы списка.

```python
list1 = [1, 3, 0, -2, 4, 5, 9, 12, -9, 16, -75, 8]

slice1 = list1[1::2]

res = 0
for i in range(len(slice1)):
    res += slice1[i]

print(res)
```

### **Задача 2**

Задан список `list2`, содержащий целые числа. Некоторые из этих чисел равны нулю. 

Задача: удалите из списка все нулевые элементы, используя инструкцию `del`.

Подсказка:

1. Используйте цикл `while`, а не `for`.
1. Подумайте, на каждой ли итерации цикла нужно инкрементировать его счётчик.

**Решение.** Поскольку для решения задачи предлагается использовать цикл while, то нам придётся определить дополнительную переменную для индекса элемента — elem_index.

```python
list2 = [1, 17, 0, 3, 6, 9, 0, 1, 34, -19, 0, 0]
elem_index = 0
```

Сам цикл строится по условию «пока индекс меньше длины списка». Так мы переберём все индексы, а затем цикл завершит свою работу.

Внутри цикла очередной элемент сравнивается с нулём. Если это нуль, то он удаляется из списка с помощью инструкции del. При этом если мы удалили элемент, то следующий встанет на его место и индекс менять не нужно. В ином случае индекс увеличивается для того, чтобы перейти к следующему элементу.

```python
while elem_index < len(list2):
    if list2[elem_index] == 0:
        del list2[elem_index]
    else:
        elem_index += 1

print(list2)
```

### **Задача 3**

Дан список списков (да, такое бывает, и достаточно часто) `list_of_lists`. Другими словами, каждый элемент списка `list_of_lists` — список сам по себе. Например, `list_of_lists[0]` – это список `['a', 'b', 'c']`, `list_of_lists[1]` – `['d', 'e', 'f']` и так далее.

Ваша задача — написать код, который объединит все элементы-списки `list_of_lists` в один-единственный список abc. Напечатайте объединённый список `abc` на экране. 

**Решение (первый способ).** Достаточно обойти вложенные в `list_of_lists` списки, добавляя их в abc конкатенацией (операция «+»). В итоге мы получим список из элементов во вложенных списках.

```python
list_of_lists = [
    ['a', 'b', 'c'],
    ['d', 'e', 'f'],
    ['g', 'h', 'i'],
    ['j', 'k', 'l'],
    ['m', 'n', 'o'],
    ['p', 'q', 'r'],
    ['s', 't', 'u'],
    ['v', 'w', 'x'],
    ['y', 'z'],
]

abc = []
for elem_index in range(len(list_of_lists)):
    abc += list_of_lists[elem_index]

print(abc)
```

**Решение (второй способ).** Используем метод extend, который расширяет исходный список переданным в extend списком. В итоге мы получим список из элементов во вложенных списках.

```python
abc = []
for elem_index in range(len(list_of_lists)):
    abc.extend(list_of_lists[elem_index])

print(abc)
```

### **Задача 4**

Даны четыре списка: `list41`, `list42`, `list43` и `list44`. 

Напишите код, который определяет, является ли каждый из этих списков `палиндромом` (читается одинаково в обоих направлениях).

__Дополнение со звёздочкой.__ Подумайте над ещё одним решением: как написать код, чтобы не повторять проверки на палиндром.

**Решение.** Список является палиндромом, если он читается одинаково в обоих направлениях. Следовательно, для определения результата нам достаточно сравнить исходный список с его развёрнутой копией. Создать развёрнутую копию списка можно через слайс с указанием –1 в шаге.

```python
list41 = ['a', 'b', 'c', 'b']
if list41 == list41[::-1]:
    print('Yes')
else:
    print('No')

list42 = ['a', 'b', 'c', 'b', 'a']
if list42 == list42[::-1]:
    print('Yes')
else:
    print('No')

list43 = ['b', 'b', 'b', 'a', 'b', 'b']
if list43 == list43[::-1]:
    print('Yes')
else:
    print('No')

list44 = ['b', 'c', 'd', 'e', 'd', 'c', 'b']
if list44 == list44[::-1]:
    print('Yes')
else:
    print('No')
```

**Решение (со звёздочкой).** Чтобы не дублировать код, мы можем создать список из исходных списков, а затем обойти эти списки, проверяя каждый из них на палиндром.

```python
list41 = ['a', 'b', 'c', 'b']
list42 = ['a', 'b', 'c', 'b', 'a']
list43 = ['b', 'b', 'b', 'a', 'b', 'b']
list44 = ['b', 'c', 'd', 'e', 'd', 'c', 'b']

all_of_them = [
    list41,
    list42,
    list43,
    list44
]

for l in all_of_them:
    if l == l[::-1]:
        print('Yes')
    else:
        print('No')
```
