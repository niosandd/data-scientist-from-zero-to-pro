Программа — это алгоритмы и структуры данных. Хорошая программа работает быстро, надёжно и предсказуемо.

Алгоритм — упорядоченная совокупность действий, которые нужно выполнить, чтобы получить желаемый результат. У алгоритмов есть начало, конец и последовательность действий, которую необходимо соблюсти.

Эффективность (или сложность) алгоритмов — количественная мера, которая позволяет сравнивать алгоритмы по скорости работы и некоторым другим параметрам. 

Структура данных — форма существования данных. Если книги — это ваши данные, то полки, на которых они стоят, — структура данных. В структуру можно добавить данные, из неё можно удалить данные, в ней можно искать данные. Эти действия называются операциями над структурами данных, или алгоритмами.

Получается такая конструкция: структуры данных состоят из данных и алгоритмов и используются для создания новых алгоритмов и структур данных.
