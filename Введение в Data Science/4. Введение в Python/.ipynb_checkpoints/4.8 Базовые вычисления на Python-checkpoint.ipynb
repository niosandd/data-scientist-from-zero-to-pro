## Базовые вычисления на Python

Мы можем использовать Python как калькулятор и выполнять любые привычные нам арифметические действия.  
  
Сложение:


```python
2 + 3 
```

Вычитание:

```python
7 - 4
```

Умножение:

```python
5 * 6 * 14
```

Деление:

```python
0.5 / 3
```

Порядок выполнения действий тот же, что мы знаем со школы: сначала умножение и деление, потом сложение и вычитание. И его тоже можно менять с помощью скобок:

```python
(7 + 3) * 2
```

Для возведения в степень используется 2 символа `*`

```python
3 ** 2
```

```python
2 ** 3
```

Можем узнать, сколько зёрнышек пшеницы окажется на [шахматной доске](https://ru.wikipedia.org/wiki/%D0%97%D0%B0%D0%B4%D0%B0%D1%87%D0%B0_%D0%BE_%D0%B7%D1%91%D1%80%D0%BD%D0%B0%D1%85_%D0%BD%D0%B0_%D1%88%D0%B0%D1%85%D0%BC%D0%B0%D1%82%D0%BD%D0%BE%D0%B9_%D0%B4%D0%BE%D1%81%D0%BA%D0%B5)

```python
2 ** 64 - 1
```

Деление нацело:


```python
17 // 3
```

Сравните с обычным делением:

```python
17 / 3
```

Остаток от деления:

```python
17 % 3
```

Если число делится нацело без остатка, какой результат мы получим?

```python
10 % 5
```

Чтобы быстрее освоиться с интерфейсом и горячими клавишами, вы можете вести в ноутбуках записи и вычисления, с которыми сталкиваетесь в жизни. И вы можете оставлять комментарии прямо рядом с вычислениями вот так:

```python
((320 + 550 + 520) + ( 240 + 150 + 340 + 600) + (310 + 550 + 150)) / 3 # среднее количество калорий за день
```

В этом уроке мы наблюдали два типа данных — integer и float — целые и дробные числа (числа с плавающей точкой).   
  
В своей голове мы можем представить числа любого размера с любым количеством порядков. Но память компьютера ограничена, и каждое число не может занимать больше определенного места.  

Это приводит к тому, что дробные числа с большим количеством значащих чисел приходится немного округлять. Иногда это приводит к парадоксам. Например:

```python
0.1 + 0.2
```
