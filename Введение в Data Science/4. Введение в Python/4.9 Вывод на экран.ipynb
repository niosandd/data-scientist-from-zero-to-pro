На прошлом уроке мы научились пользоваться Python как калькулятором. Результат вычислений выводился сразу после запускаемой ячейки:

```python
7 + 11
```

Давайте посмотрим, что будет, если мы попытаемся посчитать в одной ячейке несколько выражений одновременно:

```python
2 * 2 
228 + 282
```

Выводится только последний результат.

Даже в таком простом варианте нам это неудобно: мы бы хотели видеть все результаты. Для этого в Python есть функция `print()`, которая однозначно указывает компьютеру, что её содержимое нужно вывести на экран. Чтобы воспользоваться ей, нужно написать `print`, а в скобках указать то, что вы хотите увидеть в выводе.

```python
print(5)
```

В скобках можно указывать целые выражения. На экран будет выведен результат этих вычислений:

```python
print(3 * 12)
```

Соберём несколько выражений в одной ячейке и попросим вывести результат каждого из них:

```python
print((7 + 4) * 12 / 5)
print(2 + 4)
print(5 * 2)
```

Кроме чисел и результатов вычислений с помощью функции `print()` можно выводить на экран текст. Для этого его нужно заключить в кавычки:

```python
print("Так держать!")
```

Кавычки могут быть одинарными или двойными, разницы нет:

```python
print('один')
print("два")
```

Если вам когда-нибудь придётся анализировать популярность фильмов о Джеймсе Бонде, вы столкнётесь с проблемой: внутри названий этих фильмов нередко встречаются кавычки. Если вы попробуете написать название как есть, компьютер не поймёт, какие кавычки открывающие, а какие закрывающие, и выдаст ошибку:

```python
print("Никогда не говори "Никогда"")
```

Есть несколько способов это исправить:

Способ 1. Использовать разные виды кавычек на разных уровнях вложенности. 

```python
print("Никогда не говори 'Никогда'")
```

Способ 2. Поставить знак `\` перед кавычками, которые компьютеру нужно просто напечатать. Это называется "экранировать знак".

```python
print("Никогда не говори \"Никогда\"")
```

Одной функции `print()` можно передать сразу несколько значений для печати. Достаточно просто разделить их запятыми:

```python
print(3, 5, 7)
```

```python
print('Площадь офиса', 5 * 4, 'квадратных метров')
```

По умолчанию переданные на печать части на выводе разделяются пробелом, но можно это поменять, добавив параметр `sep`, то есть separator, разделитель:

```python
print(10, 30, 45, 0, 1, sep=',')
```

```python
print(1, 4, 1809, sep='/')
```

```python
print("Когда", "закончится", "этот", "урок!?", sep=', блин, ')
```

В одной ячейке можно несколько раз вызвать функцию `print()`, передав в каждую по несколько значений и используя разные разделители:

```python
print(2, 12, 85, 0, 6, sep='-')
print("это твой номер", "номер", "номер...", sep=", ")
```
