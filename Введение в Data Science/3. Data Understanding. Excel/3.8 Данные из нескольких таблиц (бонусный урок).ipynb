Способы борьбы с пропусками:

1. Удаление (если пропусков много — мы можем потерять много ценной информации).
2. Заполнение с помощью мер среднего уровня / самых частых значений (плохой вариант при большом количестве пропусков или дисбалансе классов).
3. Поиск дополнительного источника данных.
Поместим новые данные на новый лист в файле и обратимся к ним с помощью функции вертикального просмотра — ВПР. ВПР (VLOOKUP) производит поиск по первому столбцу диапазона и возвращает значение из найденной ячейки.

Синтаксис функции: VLOOKUP(запрос; диапазон; индекс; [отсортировано]).

Аргументы:

+ запрос — критерий поиска, то, что ищем (обычно это значение или адрес ячейки, в которой хранится значение);
+ диапазон — диапазон, в первом столбце которого будет производиться поиск по запросу;
+ индекс — номер столбца (от начала диапазона), из которого нужно взять искомое значение;
+ отсортировано — указывает, отсортированы ли данные в столбце, в котором производится поиск (рекомендуемое значение — FALSE(ЛОЖЬ)).
Пример готовой функции: =VLOOKUP(39; A2:B15; 2; FALSE).

В сочетании с IF для заполнения пропуска: =IF(ISBLANK(K2); VLOOKUP(A2; partner_data2!A$2:B; 2; FALSE); K2).
