Data Understanding — это вторая фаза методологии CRISP-DM. Цель фазы Data Understanding — поиск дополнительных данных и обогащение ими существующих данных от заказчика.

В классической форме фаза состоит из четырёх шагов.

1. Сбор данных

Существует несколько типов данных:

first-party data — это данные ваших клиентов или аудитории, которые вы собрали, вы ими владеете;
second-party data — это чьи-то данные first-party, купленные непосредственно у владельцев данных;
third-party data — это данные, которые агрегируются из множества различных источников и носят коммерческий характер;
открытые данные — это данные, законно получить которые может практически любой желающий.
Способы сбора и хранения данных:

таблица — метод структурирования данных, представляет собой распределение данных по однотипным строкам и столбцам;
реляционная база данных — метод организации данных в виде связанных таблиц с использованием SQL (Structured Query Language) — языка запросов к данным;
файл — базовый объект взаимодействия с данными в компьютерах, используемый для распространения данных;
API — набор правил, следуя которым одна программа может взаимодействовать с другой с целью получить некоторые данные.
2. Описание данных

Для этого используется описательная статистика — раздел статистики, который изучает методы представления основных свойств данных.

3. Исследование данных

Выдвигаем гипотезы и отмечаем те части данных, которые потенциально наиболее полезны.

4. Проверка качества собранных данных 

Задача — найти ошибки, содержащиеся в собранных данных, и если не исправить, то уменьшить их влияние на результат моделирования.
