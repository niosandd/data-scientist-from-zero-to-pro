Описательная (дескриптивная) статистика — раздел статистики, который изучает методы представления основных свойств данных. 

Основные задачи описательной статистики:

систематизация данных;
наглядное представление данных в виде графиков и таблиц;
количественное описание данных с помощью основных статистических показателей.
Статические показатели (или меры) делятся на две большие группы: меры среднего уровня и меры разброса или рассеяния.

Мера среднего уровня — это одно число, которое описывает целое множество чисел. Данную меру удобно применять, поскольку на основе всего одного числа можно сделать относительно верные выводы.

В Excel и Google Таблицах меру среднего уровня можно получить, если выделить необходимый диапазон данных и посмотреть на значение, отображаемое в правом нижнем углу таблицы.

Кроме того, можно воспользоваться специальными функциями, встроенными в электронные таблицы.

Функция в электронных таблицах — заранее определённая формула для произведения вычислений над ячейками.

Примеры функций для расчёта мер среднего уровня:

+ AVERAGE (СРЗНАЧ) — вычисляет среднее арифметическое. 
Если изменить значения ячеек в диапазоне, указанном в функции AVERAGE, среднее значение будет пересчитываться автоматически. Среднее не всегда является лучшей мерой для описания данных, поскольку чувствительно к выбросам. 
+ MEDIAN (МЕДИАНА) — вычисляет медиану набора чисел, то есть число, которое окажется в середине этого набора, если его отсортировать по возрастанию. Медиана менее чувствительна к выбросам, чем среднее значение. 
+ MODE (МОДА) — вычисляет моду набора чисел, то есть наиболее часто встречающееся в наборе число.
