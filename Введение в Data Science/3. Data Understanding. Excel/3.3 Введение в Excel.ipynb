Excel — это универсальная программа, которая позволяет работать с различными форматами данных и анализировать их с помощью электронных таблиц.

Аналогом Excel является сервис Google Sheets (или Google Таблицы). Данный сервис имеет ряд преимуществ:

удобство и отсутствие необходимости устанавливать дополнительные программы;
бесплатное использование;
единство интерфейса.


Создание таблицы и импорт данных
При создании нового xlsx-файла в Excel или Google Таблицах перед вами появляется пустая таблица, которая состоит из строк и столбцов (или колонок). Каждой колонке соответствует буква алфавита, а строки пронумерованы, что даёт возможность адресовать ячейки. 

Некоторые функции, которые можно использовать в Excel и Google Таблицах:

+ закрепление конкретной строки (создание заголовка таблицы);
+ работа со шрифтами: изменение самого шрифта, его стиля, размера и цвета;
+ различная сортировка (по убыванию/возрастанию численных значений, по алфавиту или в обратном порядке, сортировка диапазонов значений);
+ фильтрация колонок с помощью создания собственного фильтра.
Однако функционал в Excel и Google Таблицах намного шире. Подробнее ознакомиться с ним можно на официальном сайте Microsoft.
