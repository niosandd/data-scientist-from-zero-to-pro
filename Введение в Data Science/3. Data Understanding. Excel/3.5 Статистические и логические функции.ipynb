Книга в электронных таблицах — документ формата xlsx, в котором мы работаем.

Лист — отдельная рабочая область, которая находится в книге и позволяет работать непосредственно с таблицами и отдельными их ячейками.

Одна книга может содержать множество листов. При этом из одного листа можно обращаться к данным, находящимся в другом листе.

Для того чтобы сослаться на ячейки конкретного листа, в формулах используется следующий синтаксис: имя_листа!адрес_ячеек. Например, Лист1!A2.



Заполнение пропусков
Многие модели машинного обучения не работают с данными, в которых есть пропуски, поэтому с пропусками нужно бороться. 

Функция COUNTBLANK (СЧИТАТЬПУСТОТЫ) подсчитывает количество пустых ячеек в заданном диапазоне.

Способы борьбы с пропусками:

+ Удаление (если пропусков много — мы можем потерять много ценной информации).
+ Заполнение.
Признак — свойство объекта или явления, которое можно наблюдать или измерить. 

Признаки можно разделить на два типа:

+ числовые (возраст или размер накоплений);
+ категориальные (семейный статус или уровень образования).
Для заполнения пропусков:

1. Используем IF (ЕСЛИ) — функция возвращает различные значения в зависимости от результата логической проверки.
2. Определяем, пуста ли ячейка с помощью ISBLANK (ЕПУСТО).
3. Заполняем пропуски:
+ числовые — с помощью меры среднего уровня (среднее, медиана, мода): =IF(ISBLANK(B2); AVERAGE(B$2:B); B2);
+ категориальные — самым частым значением. Находим уникальные значения в колонке с помощью UNIQUE и считаем их частоту с помощью COUNTIF. Заменяем пропуски в таблице на самое частое значение: =IF(ISBLANK(E7); "married"; E7).
