Для выполнения практики советуем использовать Google Spreadsheets, а не Excel. Это бесплатный инструмент, необходима только авторизация через Google-аккаунт.

Часто в данных у нас возникают некие связи (функциональные зависимости), с которыми нам необходимо работать. 

Сводные таблицы — это эффективный инструмент анализа данных, который заметно упрощает поиск закономерностей и тенденций. 

Преимуществом сводных таблиц является то, что они освобождают от ручной разметки таблицы и использования большого количества различных функций. Создавая сводную таблицу, мы получаем максимально наглядное представление данных в виде таблицы.



Создание сводной таблицы
1. Выбираем «Вставка» → «Создать сводную таблицу». 
2. Указываем диапазон данных, который будет обрабатывать сводная таблица (можно оставить по умолчанию всю таблицу).
3. Выбираем, что мы хотим видеть в столбцах / строках / рассчитываемых значениях сводной таблицы. Например, COUNTA (СЧЁТЗ) подсчитывает количество значений (заполненных ячеек) в наборе данных.

Создание графиков
1. Выбираем «Вставка» → «Диаграмма».
2. При необходимости меняем настройки графика: тип графика, диапазон значений, оси, параметры и другие.

Сводные таблицы и графики можно использовать одновременно. При этом изменение каких-либо параметров и функций в таблице будет сразу же отображено и на соответствующих им графиках. Это позволяет быстро посмотреть на данные под разными углами, проверить гипотезы или выдвинуть новые.
