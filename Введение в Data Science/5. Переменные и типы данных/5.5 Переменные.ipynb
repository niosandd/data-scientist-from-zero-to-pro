# Переменные

Рассмотрим задачу. 

Трое коллег, работающих в одной комнате, обнаружили, что всем им нравится SpaceX. По такому поводу они решили пойти на их сайт и вместе заказать там мерч. Вася выбрал куртку за 140 долларов, Маша выбрала футболку за 30, Петя выбрал кружку за 20. Доставка им обошлась в 35 долларов. 

Решили посчитать, сколько это будет в рублях на каждого (по курсу 73.16 рубля за доллар):


```python
140 * 73.16
```

```python
30 * 73.16
```

```python
20 * 73.16
```

```python
35 * 73.16 / 3
```

Хотели уже заказывать, но появились более срочные дела. Вернулись к делу через неделю — курс изменился, стали считать заново. Пришлось в каждой строчке изменить значение курса:

```python
140 * 74.59
```

```python
30 * 74.59
```

```python
20 * 74.59
```

```python
35 * 74.59 / 3
```

Снова что-то не сложилось и через неделю пришлось пересчитывать ещё раз. В третий раз это уже начинает раздражать. Как бы нам только один раз ввести значение курса, чтобы оно поменялось сразу везде? 

Для этого нужно создать переменную. *Переменная* — это такой контейнер для хранения информации. 

В нашем случае информация — это курс доллара. Создадим переменную: 


```python
dollar = 75.81
```

`dollar` — это имя переменной. После знака `=` указываем значение, которое хотим ей присвоить.

Эта операция называется *инициализацией переменной*. Также её называют *объявлением* или *присваиванием переменной*.

Теперь вместо того, чтобы умножать цены покупок на значение, можем умножить их на переменную:

```python
140 * dollar
```

```python
30 * dollar
```

```python
20 * dollar
```

```python
35 * dollar / 3
```

Что при этом произошло? Python увидел выражение, в котором один из множителей является именем переменной. Движок Python посмотрел, что хранится под этим именем, и подставил вместо имени значение этой переменной. А потом уже перемножил два числа и выдал нам результат.


В примере выше мы могли записать в переменную не только курс, но и цену товаров, которые мы покупаем. Например, запишем в переменную `price` цену куртки:


```python
price = 140
```

Можем перемножить две переменные между собой, чтобы получить результат:

```python
price * dollar
```

Переменные — не обязательно числа, это могут быть и строки:


```python
item = "Куртка"
```

Чтобы посмотреть, что хранится в переменной, достаточно просто запустить ячейку с её именем:

```python
dollar
```

```python
price
```

```python
item
```

С переменными можно делать всё то же, что мы делали до этого просто с данными:

Узнать тип данных с помощью функции `type()`:

```python
type(dollar)
```

```python
type(price)
```

```python
type(item)
```

Напечатать значения с помощью функции `print()`:

```python
print(price)
```

В `print()` можно передать сразу несколько переменных, а также произвести с ними вычисления:


```python
print(item, price * dollar)
```

Значения переменных можно менять:

```python
item = "Футболка"
price = 30
print(item, price * dollar)
```

А ещё можно создавать переменные, основанные на других переменных. При этом изначальная переменная не меняется:


```python
sales_price = price * 0.8
```

```python
sales_price
```

```python
price
```

Можно также изменить переменную, основываясь на её предыдущем значении:


```python
price = price + 5
```

```python
price
```

### Имена переменных

Для имён переменных существует несколько строгих правил: 

1. Можно использовать любые буквы латиницы (большие и маленькие), цифры и нижнее подчёркивание `_`. 
2. Начинаться переменные должны с букв или нижнего подчеркивания. 
3. Названия переменных чувствительны к регистру, то есть Python отличает большие буквы от маленьких. Можно создать две разные переменные `price` с большой и с маленькой буквы и сохранить в них разные значения:



```python
Price = 3300
```

```python
price
```

```python
Price
```

Однако если мы будем использовать два таких похожих имени в реальной работе, мы быстро в них запутаемся. 
4. Всегда стоит давать переменным понятные и осмысленные названия.
