# Булевы значения

В Python булевы значения — это логический данных, который принимает только два значения: `True` и `False`. Правда или ложь.

С помощью них можно проверить, верно ли какое-то утверждение. Например, верно ли, что 3 больше 5? Для сравнения мы используем операторы сравнения `>` и `<`:


```python
3 > 5
```

```python
3 < 5
```

```python
100 > 1000000
```

```python
2555 > 1
```

Когда мы сравниваем какие-то выражения, результат перестаёт быть столь очевидным. 

Вернёмся к примеру из прошлого урока с ценами и курсами и проверим, что больше: 140 долларов или 120 евро?

Зададим курс валют с помощью ввода и предположим, что 140 долларов больше:


```python
dollar = float(input("Введите курс доллара:"))
euro = float(input("Введите курс евро:"))
140 * dollar > 120 * euro
```

В выводе `True`: мы оказались правы. Но с другим курсом валют всё может оказаться наоборот:


```python
dollar = float(input("Введите курс доллара:"))
euro = float(input("Введите курс евро:"))
140 * dollar > 120 * euro
```

Похожим образом можно проверить равенство и неравенство значений. Обратите внимание: для присвоения значений переменным мы использовали знак равенства `=`, для сравнения значений используется двойной знак равенства `==`: 

```python
price = 35
```

```python
price == 120
```

Знак неравенства записывается как восклицательный знак и равно: `!=`

```python
price != 10
```

Можно также использовать операторы `>=` ("больше либо равно") и `<=` ("меньше либо равно"):

```python
price >= 35
```

```python
price <= 30
```

Строки тоже можно сравнивать. В данном случае оператор `==` будет значить, что строки полностью совпадают:

```python
item = "Футболка"
```

```python
item == "Куртка"
```

```python
item == "Футболка"
```

Кроме операторов сравнения в Python существуют также логические операторы `and`, `or` и `not`.

Оператор `and` (И) будет выдавать в выводе `True`, только если верны оба утверждения:

```python
(3 < 5) and (15 > 10)
```

```python
(25 > 7) and (3 > 5)
```

```python
(item == "Футболка") and (price > 10)
```

Оператор `or` (ИЛИ) возвращает `True`, когда хотя бы одно из утверждений верно:


```python
(14 / 2 == 7) or (3 > 5)
```

```python
(3 * 5 != 10) or (2 * 2 == 4)
```

```python
(4 > 6) or (10 > 100)
```

Оператор `not` (НЕ) возвращает `True`, если утверждение неверно:

```python
not 35 > 30
```

```python
not 99 == 100
```

Булевы значения также используются в функциях: некоторые функции в качестве результата возвращают значения `True` или `False`. 

Например, функция `isinstance()`. Мы передаем функции какой-то объект и название типа данных. Если этот объект относится к этому типу данных — функция вернёт `True`. Если объект другого типа — вернёт `False`.


```python
isinstance(25, int)
```

```python
isinstance(2.5, int)
```

```python
isinstance("Привет!", str)
```

```python
isinstance(item, int)
```

```python
print(isinstance(price, int))
```
