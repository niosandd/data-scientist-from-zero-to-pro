## Стандартные типы данных в Python

### Числа — Numbers

Числа делятся на:
* целые (integer / int)
* дробные (float)

С числами можно производить арифметические операции. В одном выражении можно использовать и целые, и дробные числа. Иногда результат будет целым числом, иногда — дробным. 

Если сложить два целых числа, на выходе также получится целое:


```python
7 + 10
```

Если сложить два дробных, получится дробное: 

```python
4.5 + 5.4
```

Если целое и дробное — получится дробное:


```python
3.5 + 6
```

У дробного всегда будет указана дробная часть, даже если она равна нулю:

```python
2.5 + 0.5
```

Чтобы узнать, к какому типу данных относится интересующее нас число, нужно использовать функцию `type()`:

```python
type(5)
```

```python
type(4.5)
```

Даже если у дробного числа дробная часть равна нулю, оно всё равно остается дробным:


```python
type(4.0)
```

### Приведение типов

Преобразовать целое число в дробное можно с помощью функции `float()`, а обратно из дробного в целое — с помощью функции `int()`:

```python
float(7)
```

```python
int(4.9)
```

Обратите внимание, что при переводе дробного в целое, Python не округляет его, а просто отбрасывает дробную часть.

### Строки — Strings

Строки или strings (str) — это набор знаков, окруженных кавычками. 

Со строками мы не можем делать то же самое, что и с числами. Например, если мы попытаемся разделить одну строку на другую, мы получим в ответ ошибку:


```python
"привет" / "друг"
```

В этом и состоит концепция типов данных: разные типы данных могут производить разные операции. 

Что можно делать со строками? 
1. Посчитать длину строки с помощью функции `len()`. 

В выводе получим количество знаков в этой строке (пробелы тоже считаются):


```python
len("Продолжаем изучать python!")
```

2. Складывать строки с помощью оператора `+`. 

В выводе получим две строки, склеенные в одну. Такое действие называется *конкатенацией*.


```python
"Складываем " + "строки!"
```

3. Умножать строки на целые числа. 

В таком случае получим строку, повторенную нужное количество раз:

```python
"Новые задания!" * 5
```

Чтобы узнать тип данных, используем уже знакомую нам функцию `type()`:

```python
type("Наша новая строка")
```

### Преобразование чисел в строки и наоборот

С помощью функции `str()` можно преобразовать целое или дробное число строку:


```python
str(45)
```

```python
str(22.4)
```

В выводе мы видим числа в кавычках: это значит, что для Python они уже не числа, а строки.

Можем убедиться, что у нас получился нужный нам тип данных с помощью функции `type()`:

```python
type(str(55.4))
```

Также мы можем сделать и обратное: взять числа, записанные как строки, и привести их к целым и дробным числам с помощью функций `int` и `float`:

```python
int('33')
```

```python
float('43.2')
```
