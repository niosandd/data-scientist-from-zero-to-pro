# Первые ошибки

На прошлом уроке мы научились преобразовывать данные из одного типа в другой. Мы умеем менять дробные числа на целые:


```python
int(77.7)
```

Переводить строки с дробными числами в дробные числа:

```python
float('45.6')
```

А строки с целыми числами — в целые числа:

```python
int('234')
```

Попробуем перевести строку, содержащую дробное число, в целое:

```python
int('44.7')
```

В выводе видим ошибку: недопустимый знак для целого числа с основой 10. Python  сообщает нам, что в целых числах десятичной системы счисления таких знаков не бывает. Функция `int()` в десятичной системе может принимать только те строки, в которых содержатся только цифры от 0 до 9, а также `+` и `-` для обозначения отрицательных и положительных чисел. 

Если попробовать передать в функцию `int()` строку с буквами, получим ту же ошибку:


```python
int('apple')
```

### Как разобраться с ошибкой?

Довольно часто выдачи Python нам недостаточно: мы хотим не только прочитать выданную нам ошибку, но узнать больше подробностей, почему она возникла. А также понять, как с ней можно справиться. Как это сделать?

1. Найти ошибку и её решение в поисковике и на [stackoverflow.com](stackoverflow.com)

2. Поискать нужную информацию в документации — специальных документах, в которых описывается, как работает язык, как действуют функции и так далее.

[Документация по функции `int()`](https://docs.python.org/3/library/functions.html#int)

Получить справку про функцию или другой объект можно, не переключаясь из ноутбука. Достаточно написать название и добавить в конце вопросительный знак:


```python
int?
```
