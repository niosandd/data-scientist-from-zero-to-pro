# Несколько строк кода в ячейке
Когда мы только начинали пользоваться Python, каждое новое вычисление писали в новой ячейке. Потом мы стали иногда писать по 2-3 строчки кода в ячейке и запускать её целиком. Чем сложнее будут становиться наши программы, тем удобнее нам будет писать помногу строчек в одной ячейке. Некоторые куски кода будут неделимы на отдельные ячейки, другие мы будем компоновать по ячейкам из удобства.

Можно сравнить ячейку кода в ноутбуке с абзацем в тексте. Несколько предложений, объединённых общей мыслью или сюжетом. В программировании — несколько строк кода, некоторая последовательность действий, выражающая одну мысль, составляющая одну функцию.

Со временем вы сами поймёте, на какие куски удобнее делить код.

Запишем следующую задачу целиком в одной ячейке. 

Предположим, вы собрались в путешествие на машине из Петербурга в Иркутск. 

Навигатор говорит, что расстояние между городами — 5900 километров. Не всегда дороги на вашем пути будут хорошими, часто вы будете проезжать через населённые пункты — предположим, что ваша средняя скорость за весь путь будет 68 километров в час. Во время путешествия вы не хотите перенапрягаться, хотите останавливаться в красивых местах погулять, поэтому в день будете проезжать 4,5 часа. Вы знаете, что ваша машина потребляет по 8,3 литра бензина на 100 километров. Средняя цена на бензин — 45 рублей 80 копеек за литр.

Напишите код, который будет рассчитывать и выводить на экран количество полных дней путешествия (последний частичный день можно не считать), количество потраченного бензина, стоимость всего потраченного бензина.


```python
# Параметры путешествия
distance = 5900          # км — общее расстояние
speed = 68               # км/ч — средняя скорость
hours_per_day = 4.5      # часов в день
consumption = 8.3        # литров на 100 км
price_per_liter = 45.80  # рублей за литр

# Расчёты
km_per_day = speed * hours_per_day
full_days = int(distance / km_per_day)          # количество полных дней (без последнего частичного)

total_fuel = (distance / 100) * consumption     # всего бензина в литрах
total_cost = total_fuel * price_per_liter       # общая стоимость бензина

# Вывод результатов
print(f"Количество полных дней путешествия: {full_days}")
print(f"Количество потраченного бензина: {total_fuel} литров")
print(f"Стоимость всего потраченного бензина: {total_cost} рублей")
```

Вот у вас получилась ячейка с довольно большим количеством строк. Для удобства навигации по коду и обсуждения с коллегами вы можете добавить нумерацию строк в ячейке.
