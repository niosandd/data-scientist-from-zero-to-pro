# Ввод данных с экрана

Представим, что мы хотим запросить у пользователя программы его имя. Для этого в Python существует функция `input()`:

```python
input()
```

При запуске кода появляется окошко, куда можно ввести любое значение. 

Любое значение будет прочитано как строка, даже если это число:

```python
input()
```

Значение из ввода можно сохранить в качестве переменной и использовать в дальнейших действиях. Сохраним имя в переменную `name`:

```python
name = input()
```

И выводим её на печать, подставляя в готовую фразу:


```python
print("Меня зовут", name)
```

Так как `input()` любые введённые символы по умолчанию воспринимает как строку то если мы хотим использовать введённые данные как число, нам нужно привести их к числовому типу данных.

Возьмём пример из предыдущего урока, введём значение курса доллара с экрана:


```python
dollar = float(input())
```

```python
140 * dollar
```

```python
30 * dollar
```

Внутри скобок функции `input()` можно оставить пояснения к тому, что именно требуется ввести пользователю:


```python
input("Напиши свое имя:")
```

В одной ячейке можно вызвать несколько функций `input()` и сразу использовать их результат.

Создадим переменную `price`, в которую сохранится первый ввод, и переменную `dollar`, в которую сохранится второй ввод. Выведем на печать результат их произведения как “стоимость”.


```python
price = int(input("введите цену "))
dollar = float(input("введите курс доллара "))
print("Стоимость", price * dollar)
```
