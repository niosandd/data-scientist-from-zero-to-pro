Бизнес-процесс — набор шагов, которые нужно сделать, чтобы создать товар или услугу.

Задачи Business Understanding: 

1. Знакомство с бизнес-процессом заказчика.
2. Формулирование бизнес-целей проекта (например, удвоить число кредитов в месяц, не повышая риски).
3. Оценка текущей ситуации:
+ Какие есть ресурсы (люди, данные и оборудование)?
+ Какие есть ограничения (качество, производительность, внешние обстоятельства)?
+ Какие есть риски и что может пойти не так?
4. Определение целей data-аналитики: 
+ Перевести цель с языка бизнеса на технический язык и превратить её в аналитическую задачу.
+ Выбрать и согласовать технические метрики для оценки качества моделей.
+ Определить критерии успеха — минимальные значения метрик, которых мы хотим достичь.

5. Подготовка плана проекта — определить сроки и ресурсы для достижения цели.
