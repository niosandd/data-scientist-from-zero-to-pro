Методы сбора требований:

1. Интервью — свободная беседа, в ходе которой необходимо двигаться от общего к частному: используйте открытые вопросы, чтобы увидеть общую картину, закрытыми подтверждайте информацию, вносите предложения через альтернативные вопросы. Советы:

Готовьтесь к интервью. 
Делайте записи.
Кратко описывайте договорённости, выявленные на встрече, и подтверждайте их у всех участников.
2. Анкетирование — подготовка списка вопросов (анкеты) и заполнение её заказчиком.

Преимущества:

Быстрее интервью.
Стоит намного дешевле.
Недостатки:

Нет возможности задать уточняющие вопросы или переспросить.
Составление анкеты не гарантирует, что вы учтёте все необходимые вопросы.
Анкеты обычно используют для того, чтобы подтвердить уже известные требования или уточнить какие-нибудь количественные параметры решения, а не для сбора сложных и неявных требований.

Совет: нежелательно включать в анкеты открытые вопросы, старайтесь использовать закрытые и альтернативные вопросы. 

3. Мозговой штурм — вид коллективного интервью, который подразумевает свободный обмен идеями.

4. Групповые чаты — есть в любом мессенджере и позволяют использовать преимущества интервью и анкетирования.

5. Комбинирование методов — позволяет добиваться лучших результатов.
