CRISP-DM (Cross-Industry Standard Process for Data Mining) — наиболее распространённая методология по исследованию данных. Она помогает определить правильную и эффективную последовательность действий с целью уложиться в довольно сжатые сроки, которые выделены на бизнес-задачи.

CRISP-DM — это обобщённое описание подходов к проектам по аналитике данных, которые применяются на практике вне зависимости от конкретных задач.

Обычно фазы выполняются именно в такой последовательности, однако CRISP-DM — итеративная методология. Это значит, что в некоторых случаях из одной фазы можно вернуться в предыдущую, чтобы получить новую информацию или уточнить ту, которая уже есть.
