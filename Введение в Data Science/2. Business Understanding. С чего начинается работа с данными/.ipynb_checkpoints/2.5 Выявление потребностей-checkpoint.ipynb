Стейкхолдеры — люди или организации, которые так или иначе заинтересованы в результатах проекта.

Выявление стейкхолдеров — важный процесс именно потому, что они заинтересованы в проекте, а значит, у них есть собственные потребности. 

Потребности бывают абсолютно разными: одни могут лежать на поверхности, а другие могут быть далеко не очевидны. Поэтому потребности нужно выявлять и детализировать. 

Выявление потребностей — это коммуникация, в ходе которой вы уточняете бизнес-процесс заказчика, определяете стейкхолдеров и роли, которые они играют. 

Цель — это компромисс между потребностями стейкхолдеров. 

Требование — конкретная реализация потребности.

Цели vs потребности vs требования:

+ Избавиться от зубной боли — это потребность.
+ Вылечить пациента быстро и правильно — это цель.
+ Рецепт с конкретными назначениями — это требование.
