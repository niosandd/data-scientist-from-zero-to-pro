Что нужно сделать
1. Ознакомьтесь с графиком роста рынка ML/AI. Мы видим, что эта область активно развивается и новые проекты появляются во множестве сфер.

2. Вспомните примеры, которые мы разобрали в модуле. Теперь попробуем сделать следующий шаг:
+ Выберите 2-3 интересующие вас предметные области (например: медицина, финансы, ритейл, транспорт, образование, спорт).
+ Для каждой области укажите:
    + какая задача стоит перед бизнесом или людьми в этой сфере (не надо указывать глобальную, достаточно маленькой конкретной);
    + как её можно решить с помощью Data Science (например, прогнозирование спроса, распознавание изображений, автоматизация процессов, персонализация).

Зачем же всё это нужно? Главная мысль: применение Data Science не может быть отвлечённой задачей, должна решаться некоторая проблема в определённой сфере. И именно это понимание надо потренировать.

Как оформить результат
Мы подготовили шаблон (https://cloud.mail.ru/public/2e3U/sjTSd21aA) таблицы с примерами, чтобы вам было проще разобраться.
Вам нужно заполнить таблицу своими 2-3 примерами.

Важно:

+ Можете воспользоваться нашим шаблоном (Excel или Google Sheets) или создать таблицу самостоятельно в любом удобном формате.
+ Задание не нужно отправлять на проверку — оно для вас. Наша цель, чтобы вы научились связывать задачи из реальной жизни с методами Data Science.

Зачем же оформлять, если не надо сдавать? Дело в том, что мысль, выраженная в документе, и мысль, которая находится у вас в голове, — это разное. Один из важнейших навыков в Data Science — уметь выражать мысли и рассуждения в работе так, чтобы их понял любой ваш коллега, даже из другой области. И этот навык тоже тренируется в задании.
