Введение в Data Science — вводный курс. Он рассчитан на слушателей, которые ещё не обладают специальными знаниями в Data Science и программировании. 

В этом курсе вы получите базовые навыки, которые необходимы всем трём специализациям в Data Science. Познакомитесь с особенностями каждой из них. Вы сможете попробовать себя в роли дата-аналитика, ML-инженера и дата-инженера, понять, что вам нравится больше, и определить дальнейший карьерный путь. Если вы уже определились с профессией, то курс поможет вам лучше понять, чем занимаются представители смежных профессий, чтобы эффективно работать в командах.

+ Курс делится на модули, каждый из которых состоит из нескольких уроков.
+ Почти после каждого урока вас ждёт практическое задание, которое поможет самостоятельно закрепить знания и навыки, полученные в уроке. В конце модуля вы выполните домашние задания, которые проверят преподаватели. 
+ В конце курса вам предстоит подготовить итоговый проект. Вас ждёт настоящая задача из индустрии, в которой вы сможете применить все знания из курса: от загрузки данных до внедрения модели. 
+ В специальном чате в Telegram вы можете общаться с однокурсниками, задавать вопросы преподавателям и получать индивидуальную обратную связь.

Структура курса соответствует фазам CRISP-DM — одной из самых распространённых методологий в исследовании данных.

Шесть фаз CRISP DM:

+ Business Understanding — выявление целей проекта и требований со стороны бизнеса. 
+ Data Understanding — сбор данных и знакомство с ними. 
+ Data Preparation — подготовка данных и формирование итогового датасета для моделирования.
+ Modeling — построение моделей машинного обучения.
+ Evaluation — оценка качества модели. 
+ Deployment — внедрение модели.
