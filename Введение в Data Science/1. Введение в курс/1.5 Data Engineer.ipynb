Data Engineer (DE) — строит надёжную и функциональную инфраструктуру для хранения, перемещения и обработки данных.

Компетенции дата-инженера:

ETL — сбор, преобразование и загрузка данных (extract, transform, load).

Пример работы дата-инженера:

Матч Ли Седоля против AlphaGo.
Матч проводился в Сеуле, а программа AlfaGo исполнялась в вычислительном кластере из тысяч микропроцессоров и сотен графических ускорителей. Google пришлось прокладывать оптоволоконную линию связи, чтобы обеспечить высокоскоростное сетевое соединение с кластером AlphaGo. 
Дата-инженеры создают в компаниях те условия, благодаря которым дата-аналитики получают необходимые для анализа данные, а ML-инженеры учат свои модели и превращают их в успешные продукты.
