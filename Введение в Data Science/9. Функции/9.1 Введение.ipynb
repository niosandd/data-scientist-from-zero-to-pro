# Функции. Введение

Рассмотрим программу, которая считает среднюю сумму трат на продукты в сервисе доставки. 

Например, вы сделали 5 заказов: 3 их них на 1000 рублей, остальные – на 1200. Ваш средний чек (то есть средняя сумма трат) – это 1080 рублей. По сумме среднего чека обычно анализируют покупательскую способность клиента – сколько средств он может и готов потратить на продукты.

Сама программа выглядит вот таким образом. Ничего сложного: считаем сумму всех покупок в цикле, обновляя переменную `overall_sum`, и делим на количество покупок, получаем результат. В данном случае – 1180 рублей.



```python
user_1 = [1200, 1000, 1000, 1000, 1200]
```

```python
overall_sum = 0 
for i in range(len(user_1)):
    overall_sum += user_1[i]

print(overall_sum/len(user_1)) 
```

Как только нам будет необходимо посчитать средний чек для другого клиента, суммы чеков которого лежат в списке `user_2`, нам придётся:
- скопировать этот код
- изменить везде список `user_1` на `user_2`  
- и снова его запустить




```python
user_2 = [200, 100, 1000, 300, 500, 1200, 1000]
```

```python
overall_sum = 0 
for i in range(len(user_2)):
    overall_sum += user_2[i]

print(overall_sum/len(user_2)) 
```

Чем больше средних чеков нам нужно будет посчитать, тем сильнее будет разрастаться наш код. В конце концов, в нём будет очень сложно ориентироваться и нам, как дата-сайентистам, и нашим коллегам, которые хотели бы посмотреть результаты наших исследований. 
Было бы намного удобнее, если бы можно было вызвать некоторую команду, скажем, `get_average`, которая в одну строчку считает среднее на указанном списке:

```python
#@title
#@Код функции get_average
def get_average(user_1):
    overall_sum = 0 
    for i in range(len(user_1)):
        overall_sum += user_1[i]

    print(overall_sum/len(user_1))  
```

```python
user_1 = [1200, 1000, 1000, 1000, 1200]
user_2 = [200, 100, 1000, 300, 500, 1200, 1000]
get_average(user_1)
get_average(user_2)
```

4 ячейки, которые занимали 10 строк кода, сузились до 1 ячейки и 4 строк кода. Более того – код стал намного читабельнее. 
Если ноутбук будет состоять из таких понятных команд, как `get_average`, его без труда сможет понять и использовать ваш коллега. А это очень важно в работе: любое исследование или разработка должны быть  понятны коллегам, чтобы они могли с ними работать. 

Такие команды как `get_average` в Python называются ***функциями***. Функции – это один из механизмов реализации принципа хорошего кода *DRY*, что значит *don’t repeat yourself*. Написав какой-то кусок кода – не пиши его ещё раз.
