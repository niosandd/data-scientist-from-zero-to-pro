###  Числа Фибоначчи

Фибоначчи — это математик из средневековой Европы, представивший в 1202 году в своей книге последовательность чисел, которая в будущем стала называться последовательностью Фибоначчи.

Он поставил теоретическую задачу о развитии популяции, которая звучала следующим образом: 

Пусть есть пара кроликов, один самец и одна самка в поле. Сколько пар кроликов они произведут через год при следующих условиях:
- Ни один кролик не умирает и не съедается хищниками
- Каждая самка размножается каждый месяц, начиная со второго месяца её жизни
- Каждый раз, когда самка размножается, она рожает одну пару кроликов: одного самца и одну самку.

Тогда схема развития популяции кроликов будет выглядеть следующим образом: 
- Изначально была одна пара в январе. 
- В феврале по условию задачи они ещё не могут произвести потомство.
- А вот уже в марте месяце их стало две пары.
- Эта новая пара через два месяца стала производить потомство и в 5 месяце их уже 5 пар.

Число пар, начиная с марта, можно рассчитать как сумму пар за два предыдущих месяца. В марте число 2 — это сумма из 1 пары в феврале и 1 пары в январе. В мае — это сумма пар апреля (3) и марта(2). В этом и есть закономерность чисел Фибоначчи.

Нулевой элемент последовательности Фибоначчи — 1, первый – 1, второй — 2, третий — 3, чевертый — 5 и так далее.


Запишем формулу для расчёта чисел Фибоначчи:

пусть `f` — функция расчёта числа Фибоначчи, `n` — это номер числа Фибоначчи в последовательности, `f(n)` — это число Фибоначчи, находящееся в последовательности под номером `n`.

Число Фибоначчи на месте `n` — это сумма чисел Фибоначчи на двух предыдущих местах: `n-1` и `n-2`. Это условие работает только начиная со второго элемента последовательности:

`f(n) = f(n-1) + f(n-2), n>=2`

Поскольку первый и нулевой элементы последовательности мы по этой формуле рассчитать не можем, явно запишем, что они равны 1:

`f(1) = 1, f(0) = 1`

Объявим функцию `fibonacci`, которая на вход принимает один аргумент `n` —  номер числа в последовательности, которое мы хотим получить. Если подставим `n = 3`, должны получить 3. Если `n = 4` — 5.

Мы знаем, что если `n` равно 0 или 1, то значение числа Фибоначчи — 1, запишем это условие.


````python
def fibonacci(n):
    if (n == 0) or (n == 1):
        return 1
````

Если же `n` не 1 и не 0, то нам нужно два предыдущих числа Фибоначчи. В `else` запишем возвращаемое значение как сумму предыдущего — `fibonacci(n-1)` — и предпредыдущего — `fibonacci(n-2)`. Вот мы и вызвали функцию внутри самой себя, то есть применили рекурсию.




````python
def fibonacci(n):
    if (n == 0) or (n == 1):
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

Распишем, как будет вести себя программа при `n = 3`:
1. `n` не равно 1 и 0, значит — уходим в `else`. 
2. Функция возвращает сумму значений `fibonacci(n-1)` и `fibonacci(n-2)`. 
3. Запускается функция `fibonacci(n-1)`. `n-1` равно 2, 2 не равно 1 и 0, переходим в `else` — значение равно `fibonacci(1) + fibonacci(0)`.
4. Запускается функция `fibonacci(1)`. Возвращается результат 1 и подставляется в выражение вместо `fibonacci(1)`.
5. Далее считается  `fibonacci(0)`. Возвращается результат 1 и подставляется в выражение вместо `fibonacci(0)`.
6. Итого выражение для `n=2` полностью посчиталось, `fibonacci(2)` равно 2.
7. Возвращаемся к ещё не рассчитанной функции `fibonacci(n-2)`. `3 - 2 = 1`, значит надо посчитать Фибоначчи от 1. Это 1. Подставляем в результат.
8. Итого Фибоначчи от 3 — это `2 + 1 = 3`.

В рекурсивной функции есть две главные вещи: 
- условие для остановки
- условие для продолжения расчёта

В нашем случае условие остановки — это первый `if`. Если условие в нём выполнится, то вызов функции самой себя не произойдёт и она вернёт число.
Условие для продолжения расчёта — это `else`. Если `n` больше 1, то мы продолжаем считать и функция вызывает саму себя.

Без критерия остановки рекурсивная функция никогда не прекратится, будет вызывать саму себя бесконечно, оперативная память компьютера заполнится и, скорее всего, придётся перезапускать компьютер. С рекурсией нужно быть осторожным.

Часто некоторые задачи можно решить с помощью рекурсии, а можно и без, но сложность кода при этом сильно возрастает. Под сложностью мы подразумеваем количество операций, циклов и условий, которые придётся использовать. С другой стороны, рекурсия любит покушать память. Так что придётся делать выбор: какой алгоритм выгоднее реализовать. 

### Задача вычисления степени с помощью рекурсии

Ещё один показательный пример задачи, которую можно решить с помощью рекурсии, это вычисление степени двойки. Представьте, что вам нужно вычислить значение 2 в степени n.
- чтобы найти 2 в степени 3 — нужно 2 в степени 2 умножить на 2
- чтобы найти 2 в степени 2 — нужно 2 в степени 1 умножить на 2
- чтобы найти 2 в степени 1 — нужно 2 в степени 0 умножить на 2
- 2 в степени 0 — 1.

Запишем рекурсивную формулу: 

`2^n = 2 * 2^(n-1), n >= 1`

`2^0 = 1`

`2^0 = 1` — это критерий остановки, а `2 * 2^(n-1)` — это условие продолжения.

Запрограммируем функцию для вычисления степени: 

````python
def power_2(n):
    if n != 0:
        return 2 * power_2(n-1)
    else:
        return 1
````

Вот такая функция получилась. Проверим на нескольких примерах:

````python
power_2(3)
````

````python
power_2(5)
````

````python
power_2(0)
````

### Рекурсия или цикл?

Многие задачи можно решить и через рекурсию, и через цикл. Вот как выбрать подходящий инструмент:

| Ситуация | Рекурсия | Цикл |
|---|---|---|
| Задача сама по себе рекурсивна (дерево, вложенные структуры, Фибоначчи) | ✅ код короче и читаемее | ⚠️ громоздко, много вложений |
| Простой перебор списка или диапазона | ⚠️ лишняя сложность | ✅ `for`/`while` прямолинейны |
| Глубина вложенности > 1000 шагов | ❌ Python упадёт с `RecursionError` | ✅ ограничений нет |
| Критична скорость и память | ⚠️ каждый вызов создаёт новый фрейм в памяти | ✅ работает быстрее |

**Правило:** если задача формулируется через саму себя — рекурсия. Если нужно просто пройтись по элементам — цикл.

---

**Пример: `fibonacci` через цикл и через рекурсию**

```python
# Цикл — быстрее, не ест стек, но менее наглядно
def fibonacci_loop(n):
    a, b = 1, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# Рекурсия — точно повторяет математическое определение
def fibonacci(n):
    if n == 0 or n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
```

Оба варианта дают одинаковый результат. Рекурсивный проще читается, но при больших `n` (больше 35–40) заметно замедляется: одни и те же значения пересчитываются по многу раз. Итеративный вариант лишён этого недостатка.
