## 1. Функции

### Объявление и вызов

```python
def имя_функции(параметр1, параметр2):
    # тело функции
    return результат
```

- `def` — начало объявления
- Параметры — локальные переменные, видны только внутри функции
- Функцию можно вызвать сколько угодно раз с разными аргументами

### return

`return` возвращает значение из функции и **немедленно завершает** её работу. Весь код после `return` не выполняется.

````python
def square(x):
    return x ** 2
    print('сюда не дойдём')  # никогда не выполнится

result = square(5)
print(result)   # 25
````

**Без `return` функция возвращает `None`:**

````python
def greet(name):
    print(f'Привет, {name}!')  # выводит на экран, но ничего не возвращает

saved = greet('Аня')
print(saved)   # None — сохранить не получилось
````

> **Правило:** если результат нужно использовать дальше — пиши `return`. Если только вывести на экран — `print()` внутри функции.

## 2. Именованные аргументы и параметры по умолчанию

### Именованные аргументы

При вызове можно явно указать, какое значение в какой параметр идёт. Порядок перестаёт быть важным:

````python
def convert(amount, currency):
    pass

convert(7200, 'USD')                    # обычный вызов — порядок важен
convert(currency='USD', amount=7200)    # именованные — порядок не важен
````

### Параметры по умолчанию

В объявлении функции можно задать значение по умолчанию через `=`. Тогда аргумент можно не передавать — Python подставит дефолт сам:

````python
def convert(amount, currency='USD'):  # currency по умолчанию 'USD'
    pass

convert(7200)          # currency = 'USD'  (берётся дефолт)
convert(7200, 'CNY')   # currency = 'CNY'  (дефолт перезаписан)
````

> Параметры по умолчанию всегда идут **после** обязательных параметров.

## 3. Декомпозиция и встроенные функции

### Декомпозиция

Сложную задачу разбивают на маленькие функции — каждая делает одно действие. Принцип **DRY** (Don't Repeat Yourself): не писать один и тот же код дважды.

````python
def rub_to_usd(amount):
    return amount / 70.99

def usd_to_cny(amount):
    return amount * 6.44

def rub_to_cny(amount):          # составная функция из двух маленьких
    usd = rub_to_usd(amount)
    return usd_to_cny(usd)
````

### Встроенные функции

| Функция | Что делает |
|---|---|
| `len(lst)` | длина списка |
| `sum(lst)` | сумма элементов |
| `min(lst)` | минимум |
| `max(lst)` | максимум |

````python
scores = [4, 5, 3, 2, 5]
print(sum(scores) / len(scores))  # среднее → 3.8
print(max(scores))                # 5
print(min(scores))                # 2
````

### Список списков

Список, каждый элемент которого — тоже список. Внешний цикл идёт по подспискам, внутренний — по элементам внутри:

````python
scores = [
    [4, 5, 3],   # класс 1
    [2, 3, 4],   # класс 2
]

for classroom in scores:
    print(sum(classroom) / len(classroom))
````

Чтобы объединить список списков в один плоский список:

````python
def extend_list(list_of_lists):
    result = []
    for lst in list_of_lists:
        result += lst      # += для списков — «склеивание»
    return result

# До:  [[1, 2], [3, 4]]
# После: [1, 2, 3, 4]
extend_list([[1, 2], [3, 4]])
````

## 4. Строки — методы

### Регистр

````python
'МоСкВа'.lower()   # 'москва'
'москва'.upper()   # 'МОСКВА'
````

### split() → возвращает список

`split()` разбивает строку по разделителю и **возвращает список**. К нему сразу применяются индексы `[0]`, `[1]`, `[2]`:

````python
fio = 'Иванов Герасим Александрович'
parts = fio.split()   # ['Иванов', 'Герасим', 'Александрович']

parts[0]   # 'Иванов'        — фамилия
parts[1]   # 'Герасим'       — имя
parts[2]   # 'Александрович' — отчество
````

Можно указать свой разделитель:

````python
'name=Аня,phone=88002,city=москва'.split(',')
# → ['name=Аня', 'phone=88002', 'city=москва']

'name=Аня'.split('=')
# → ['name', 'Аня']
````

### find()

Ищет подстроку и возвращает индекс начала. Если не найдена — возвращает `-1`:

````python
'город москва'.find('москва')   # 6
'город москва'.find('казань')   # -1

# Применение: проверить наличие слова
text = 'г москва'
if text.find('москва') != -1:
    print('Москва найдена')
````

### replace()

Заменяет все вхождения одной подстроки на другую:

````python
'8800 234 56'.replace(' ', '')   # '880023456' — убираем пробелы
'!!!hello'.replace('!', '')      # 'hello'
````

### Строка как список: индексы и слайсинг

````python
s = '??Москва'
s[0]    # '?'       — первый символ
s[-1]   # 'а'       — последний символ
s[2:]   # 'Москва'  — всё с индекса 2 до конца
s[:2]   # '??'      — первые два символа
s[1:-1] # '?Москв'  — без первого и последнего
````

### f-строки

````python
name = 'Аня'
city = 'Москва'
print(f'Имя: {name}, Город: {city}')  # Имя: Аня, Город: Москва
print(f'Сумма: {2 + 3}')              # Сумма: 5
````

### Цепочка методов

Методы применяются слева направо — результат каждого передаётся следующему:

````python
'  МоСкВа  '.strip().lower()   # 'москва'
# strip() убирает пробелы по краям → lower() приводит к нижнему регистру

'МоСкВа'.lower().split()       # ['москва']
# lower() → 'москва'  →  split() → ['москва']
````

## 5. Рекурсия

**Рекурсивная функция** — функция, которая вызывает сама себя.

### Два обязательных элемента

1. **Условие остановки** — базовый случай: функция возвращает результат и НЕ вызывает себя снова
2. **Условие продолжения** — вызов себя с изменённым аргументом, который шаг за шагом ведёт к остановке

````python
def factorial(n):
    if n == 0:                    # условие остановки: 0! = 1
        return 1
    else:
        return n * factorial(n-1) # условие продолжения

factorial(4)   # 4 * 3 * 2 * 1 = 24
````

### Как разворачивается factorial(3)

```
factorial(3)
  → 3 * factorial(2)
       → 2 * factorial(1)
              → 1 * factorial(0)
                     → 1   ← условие остановки достигнуто
```

### Палиндром — подсказка к логике

Слово — палиндром, если:
- оно пустое или из одной буквы → **True** (условие остановки)
- первая и последняя буквы совпадают **и** середина тоже палиндром → **True**
- иначе → **False**

````python
# Полезные операции со строкой для задачи на палиндром:
s = 'abcba'
print(s[0])     # 'a'    — первый символ
print(s[-1])    # 'a'    — последний символ
print(s[1:-1])  # 'bcb'  — середина (без первого и последнего)
print(len(s))   # 5      — длина строки
````

### Рекурсия vs цикл

| Ситуация | Рекурсия | Цикл |
|---|---|---|
| Задача сама по себе рекурсивна (факториал, палиндром) | ✅ код читаемее | ⚠️ громоздко |
| Простой перебор списка | ⚠️ лишнее | ✅ `for`/`while` |
| Глубина > 1000 шагов | ❌ `RecursionError` | ✅ без ограничений |
