## Цикл while

Когда мы рассматривали цикл `for`, количество циклов, которые мы запускали, было строго определено – либо всеми значениями, которые мы перечислили, либо диапазоном, который мы сгенерировали. 

Цикл `while` действует по-другому. Само слово `while` в этом контексте означает “до тех пор”, “покуда”. И цикл строится в соответствии с этим значением:

Покуда выполняется условие, которое мы зададим, цикл будет повторяться. Если условие перестанет выполняться, цикл завершится. 

Рассмотрим это на примере про банковский вклад из прошлого урока. Предположим, что мы не хотим создавать вклад на какое-то определенное количество лет, а просто хотим из наших 139 тысяч получить 220 тысяч. Когда эта сумма накопится, мы забираем деньги из банка. Поставим эту границу как условие для цикла `while`. И добавим фразу после цикла.


```python
money = 139990
while money < 220000:
    money *= 1.09
    print("На счету будет", money, "рублей")
print("Я забираю деньги!")
```

Python вывел 6 строк и остановился. Рассмотрим, как исполнялся код:
- Python начал исполнять цикл и сравнил значение `money` с 220 тысячами. 
- Условие было выполнено, поэтому он пошел исполнять блок кода внутри цикла. 
- Поменял значение переменной и вывел на печать. 
- Пошел на второй круг, снова проверил условие – то есть сравнил новое значение переменной `money` с 220 тысячами
- Условие слова было выполнено, и поэтому он снова стал исполнять блок внутри цикла. И так при каждом новом цикле. 
- На шестом цикле значение переменной из 215 тысяч стало 234 тысячами и было выведено на печать. 
- Python пошёл на седьмой круг, сравнил 234 тысячи с 220 и обнаружил, что условие не выполняется. 
- После этого он пропустил всё, что написано внутри цикла, и перешёл к исполнению следующего за циклом блока: вывел фразу о том, что мы забираем деньги.

Если мы изменим значение в условии, изменится и количество повторений.
В нашем примере мы написали нашу фразу после цикла, но мы также могли использовать команду `else`, аналогично тому, как мы использовали её в конструкции `if else`. 


```python
money = 139990
while money < 220000:
    money *= 1.09
    print("На счету будет", money, "рублей")
else:
    print("Я забираю деньги!")
```

Ограничение по условию очень удобно, но таит в себе одну опасность. Если вы зададите условие, которое выполняется всегда, ваш цикл не сможет остановиться. 

Например, если мы случайно опечатались в знаке и в количестве нулей и задали границу не меньше 220, а больше 22 тысяч. Переменная `money` у нас была изначально больше 22 тысяч, а с каждым циклом только увеличивается. Цикл становится бесконечным, прервать его можно только вручную, остановив исполнение кода.


```python
money = 139990
while money > 22000:
    money *= 1.09
    print("На счету будет", money, "рублей")
else:
    print("Я забираю деньги!")
```

Как спасти положение? Мы можем добавить условие, при котором цикл прервётся в любом случае. Например, ограничить сумму в банке той, что может быть возмещена по страховке – сейчас это 1 миллион 400 тысяч. Если количество денег становится больше, мы останавливаем цикл. Добавим это условие. Выведем на печать намерение остановить цикл и добавим команду `break` – прерывание.


```python
money = 139990
while money > 22000:
    money *= 1.09
    print("На счету будет", money, "рублей")
    if money >= 1400000:
        print("Останавливаем цикл")
        break
else:
    print("Я забираю деньги!")
```

Цикл сделал больше десятка повторов и не собирался останавливаться по условию `while` – оно всегда выполнялось. Зато прервался по нашему дополнительному условию. Заметьте, что указанное в части `else` не было выполнено вообще,поскольку не было случая, когда условие `while` не было бы выполнено, чтобы перейти к части “в противном случае”.

В некоторых случаях нам удобно специально сделать цикл бесконечным и прерывать его только при исполнении внутреннего условия. Например, мы можем создать программу, которая будет бесконечно задавать пользователю какой-то вопрос, пока он не прикажет ей остановиться. Чтобы условие `while` всегда выполнялось, мы можем просто напрямую и написать `while True`. Запросить пользователя что-то ввести, чтобы это бесконечно спрашивалось. Но добавить условие:


```python
while True:
    word = input("Скажи что-нибудь, а?")
    if word == "отстань":
        break
```

Программа готова повторять цикл бесконечно, пока мы не прервем его.
