## Цикл for

Представим такую ситуацию: вы решили порадовать себя iPhone последней модели. Но отдать за покупку разом 139 990 рублей вам жалко. Вы рассматриваете вариант рассрочки и хотите понять, на сколько месяцев надо брать рассрочку, чтобы и размер выплаты был комфортным, и чтобы не затягивать надолго с оплатой телефона. Чтобы прикинуть, вы рассчитываете в Python, какой будет размер месячной оплаты при рассрочке на 3 месяца, 6, 9 и 12. 

Сначала напишем для трёх месяцев:


```python
price = 139990
months = 3
print("При рассрочке на", months,"мес. - ", price / months)
```

А теперь скопируем и поменяем количество месяцев для остальных вариантов:


```python
price = 139990
months = 6
print("При рассрочке на", months,"мес. - ", price / months)
```

```python
price = 139990
months = 9
print("При рассрочке на", months,"мес. - ", price / months)
```

```python
price = 139990
months = 12
print("При рассрочке на", months,"мес. - ", price / months)
```

Отлично получилось. Но пришлось много копировать и менять вручную. И весь код кроме одной переменной повторяется, это выглядит неоптимально. 

Мы могли бы не копировать, а только менять переменную и перезапускать ячейку.
Но в таком случае мы видим на печати только результат последнего запуска. Это тоже не очень удобно, не хранить же нам всё это в собственной памяти?

Хотелось бы как-то перезапускать код с разными значениями, но и так, чтобы на выводе были все результаты.
Такая возможность в Python есть благодаря циклам. По-английски это называется loop – петля. Напишем цикл.

- Начинаем с ключевого слова `for`. 
- Указываем переменную, которая будет меняться – в нашем случае месяцы, `months`. 
- После этого идёт ключевое слово `in`, в данном контексте оно значит “среди”, “среди следующих значений”. 
- Через запятую перечисляем значения, которые мы хотим поочерёдно придать этой переменной. 
- Ставим двоеточие, начинаем новый блок. В нем у нас будет код, который мы уже исполняли, он вычисляет рассрочку. Переменную `price` мы не трогали, она осталась с прежних времен.


```python
for months in 3, 6, 9, 12:
    print("При рассрочке на", months,"мес. - ", price / months)
```

Как исполнялся этот код?

- Первый раз считывая цикл, Python обнаружил переменную `months` и присвоил ей первое значение из ряда, в данном случае 3. 
- После этого он перешел к блоку, который стоит после двоеточия, и стал исполнять его с этим значением переменной. 
- Всё посчитал и вывел на экран – получилась первая строка вывода. 
- Посмотрел дальше, но больше инструкций в этом блоке не увидел.
- Пошёл по второму кругу цикла – вернулся к переменной `months` и теперь присвоил ей следующее значение, то есть 6. 
- Исполнил инструкции после двоеточия с этим новым значением переменной. 
- Пошел на новый цикл, с 9. И делал так, пока не перебрал все значения, которые мы указали. 
- Обнаружив, что новых значений нет, Python пошел искать инструкции после цикла, но мы ему таких не написали, поэтому он закончил программу.




Немного усложним программу, чтобы лучше понять её ход – допишем строки до цикла, внутри цикла, и после цикла. Теперь цену мы будем запрашивать с помощью `input()`, чтобы можно было проверять для разных вариантов. В цикле мы добавим наше оценочное суждение, которое не использует переменные. В конце добавим заключительную фразу:


```python
price = int(input("Сколько стоит обновка?"))
for months in 3, 6, 9, 12:
    print("При рассрочке на", months,"мес. - ", price / months)
    print("Не дороговато ли?")
print("Теперь есть из чего выбрать")
```

Первая строка исполнилась 1 раз: спросила нас о стоимости.

Дальше две инструкции исполнились 4 раза. 

Последняя инструкция после цикла исполнилась снова 1 раз.

Только то, что стоит единым блоком с отступом, то есть блок цикла, закольцовано и выполняется четыре раза, остальное выполняется как обычно один раз.



Мы посмотрели на эти 4 варианта и решили, что нам этого мало: мы хотим увидеть все варианты, для любого количества месяцев – от минимума до 2 лет.

Но печатать все эти значения мы не хотим я – будем использовать функцию `range()`. Это специальная функция, которая генерирует последовательный набор чисел. 

Она принимает три аргумента: 
1. Первый аргумент указывает, с какого числа начать. Рассрочка в один месяц – это значит заплатить сразу, это и так понятно. Поэтому начнём с двух месяцев. 
2. Второй аргумент – на каком числе остановиться. Важно, что само число, которое мы укажем, уже не будет входить в диапазон. Так что если мы хотим остановиться на 24 месяцах, в качестве границы нам нужно указать следующий месяц – двадцать пятый. 
3. Третий аргумент – это шаг, с которым будут увеличиваться числа. Мы хотели посмотреть на все месяцы, так что поставим шаг равным единице.

```python
price = int(input("Сколько стоит обновка?"))
for months in range(2, 25, 1):
    print("При рассрочке на", months,"мес. - ", price / months)
    print("Не дороговато ли?")
print("Теперь есть из чего выбрать")
```

Итак, у нас всё те же несколько строк кода, а мы сгенерировали расчёт для 23 вариантов.

С такой же легкостью мы могли бы посчитать, например, для поквартального расчета до пяти лет. Поставим начало на трёх месяцах, верхнюю границу даже считать не будем, а поручим это Python, умножив количество месяцев в году на пять и добавив единицу (чтобы включить в границы все пять лет), а шаг поставим 3 месяца – 1 квартал:


```python
price = int(input("Сколько стоит обновка?"))
for months in range(3, 12 * 5 + 1, 3):
    print("При рассрочке на", months,"мес. - ", price / months)
    print("Не дороговато ли?")
print("Теперь есть из чего выбрать")
```

### Функция `range()`

Если просто запутсить функцию `range() отдельно, мы не увидим на выходе последовательность. Вместо неё – просто условия заданного диапазона. 


```python
range(2, 12, 2)
```

Если напечатать – то же самое:

```python
print(range(2, 12, 2))
```

Если запросить тип – увидите тип `range`:

```python
type(range(2, 12, 2))
```

Функция `range()` генерирует последовательность чисел, но чтобы не занимать лишние ресурсы, она не делает это заранее, сохраняя где-то список чисел, а выдаёт каждое следующее, только когда оно требуется. 

Так что чтобы потребовать от функции каждое из чисел и получить его, нам понадобится цикл. Напишем маленький цикл, который просто выведет нам по очереди все числа из диапазона на печать:


```python
for i in range(2, 12, 2):
    print(i)
```

Мы задали начало, конец и шаг. Но у этой функции есть значения аргументов по умолчанию. Для шага это значение 1. Так что если мы хотим вывести все числа подряд – можем просто убрать последний аргумент, оставив начало и конец диапазона:


```python
for i in range(2, 12):
    print(i)
```

Теперь Python вывел все числа подряд.

Для начала диапазона значение по умолчанию – ноль. Уберём первый аргумент и посмотрим:


```python
for i in range(12):
    print(i)
```

Теперь отсчёт начался с нуля и шел с шагом один. Последний аргумент – конец диапазона – убирать нельзя, без него функция не сработает.

Вернёмся к циклам. Представим, что вы повздыхали над новым айфоном, но решили, что это слишком дорого, лучше вы пока что будете копить деньги. И понесли накопленное на айфон в банк, положить под проценты. Например, на 5 лет. 

Вам обещают 9 процентов годовых с выплатой раз в год. Вы хотите узнать, сколько это будет через пять лет. Но поскольку вы не собираетесь забирать из банка доход от процентов, а будете оставлять его на счету, вам нужно посчитать сложные проценты за пять лет - процент от основного вклада и от его процентов. 

Можно сделать это с помощью циклов. Нам нужен цикл на 5 раз (то есть пять лет), который будет брать количество денег на счету и умножать их на 1.09, то есть получать сумму с процентами. В следующий проход этот цикл должен брать уже получившуюся сумму и снова её умножать на 1.09. 

Напишем такой цикл:
- в переменной `money` зададим исходное количество денег
- укажем, что цикл нужно провести пять раз. Для этого используем знакомую функцию range(). 
- внутри цикла укажем нужную арифметику: переменной будем присваивать новое значение на основании предыдущего значения. 
- в конце выведем на печать результат.


```python
money = 139990
for i in range(5):
    money = money * 1.09
    print("На счету будет", money, "рублей")
```

- Python запомнил, что переменная `money` равна цене айфона
- Начал считывать цикл и присвоил переменной `i` её первое значение из диапазона – 0.
- Пересчитал переменную `money`, умножив её изначальное значение на 1.09
- Вывел на печать
- Зашел на второй круг: переменная `i` стала 1, а переменная `money` снова была пересчитана. Но мы уже умножали не 139 990, а 152 589.1, накопленные за год благодаря процентам. 
- Python пошёл на следующий круг и дальше на следующий, перебирая все значения `i`: 2, 3, 4. И всё, пятеркой `i` уже стать не должно: диапазон закончился и остановилась наша программа.

В этом случае мы не использовали переменную `i`, которая определяла цикл, в своих расчетах процентов. Она нам нужна была только как счётчик, чтобы отсчитать только 5 лет. 

 

Рассмотрим еще один пример. 

За окном осень. Ночи становятся всё длиннее, а температура на улице всё ниже. Можем приблизительно установить, что с каждой неделей ночь становится длиннее на полчаса, а температура ниже на полтора градуса. Давайте посчитаем, как она будет меняться по неделям в ближайшие 10 недель. 

Зададим текущую температуру и продолжительность ночи. Пишем цикл по неделям в диапазоне от 1 до 10 (ставим 11, потому что хотим включить десятую неделю). Внутри цикла укажем расчёты и выведем результаты на экран, с указанием номера недели.


```python
temperature = 10
night = 11.5
for week in range(1, 11):
    temperature = temperature - 1.5
    night = night + 0.5
    print("Неделя", week)
    print("Температура", temperature, "градусов Цельсия")
    print("Продолжительность ночи", night, "часов")
```

Переменная `week` просто ведет отсчёт. Другая переменная постепенно уменьшается, третья постепенно увеличивается. 

В программах довольно часто встречается ситуация, когда в цикле мы высчитываем значение переменной, исходя из её прошлого значения. Поэтому для такого действия есть более короткая запись. Вместо написания имени переменной, мы можем поставить оператор `+=` или `-=` и указать значение, на которое хотим увеличить или уменьшить переменную:


```python
temperature = 10
night = 11.5
for week in range(1, 11):
    temperature -= 1.5
    night += 0.5
    print("Неделя", week)
    print("Температура", temperature, "градусов Цельсия")
    print("Продолжительность ночи", night, "часов")
```

Результат получился тот же, что и в прошлый раз.

Кроме `+=` и `-=` мы также можем использовать `*=` (когда вычисляем переменную, умножаем её же на какое-то число) или `/=` (деление равно).
